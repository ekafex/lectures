# **Nga Fizika Tek Elektronika**  

Elektronika buron nga **elektromagnetizmi klasik**, i cili përshkruhet nga **ekuacionet e Maxwell-it**. Këto ekuacione përshkruajnë sjelljen e fushave elektrike dhe magnetike, si dhe ndërveprimin e tyre me ngarkesat dhe rrymat elektrike.  

Në formën diferenciale, ekuacionet e Maxwell-it janë:  

$$
\begin{array}{rl} 
\nabla \times \mathbf{E} &= -\frac{\partial \mathbf{B}}{\partial t}\\
\nabla \times \mathbf{B} &= \mu_0 \mathbf{J} + \mu_0 \varepsilon_0 \frac{\partial \mathbf{E}}{\partial t}\\
\nabla \cdot \mathbf{E} &= \frac{\rho}{\varepsilon_0}\\
\nabla \cdot \mathbf{B} &= 0
\end{array}
$$

Këto ekuacione janë themelore për përshkrimin e të gjitha fenomeneve elektromagnetike, por zgjidhja e tyre për sisteme reale është komplekse. Këtu hyn **Disiplina e Materies së Përqendruar në formë Kompakte (Lumped Matter Discipline - LMD)**, një qasje që lehtëson analizën e qarqeve elektrike.


### **Disiplina e Materies së Përqendruar në formë Kompakte (LMD)**
LMD është një paraqitje e thjeshtëzuar e një sistemi fizik apo qarku duke supozuar që elementët e tij janë të përqëndruar në një pikë dhe sjellja e tyre mund të përshkruhet nga një model matematikor i idealizuar. LMD thjeshtëzon qarkun elektrik real në një graf elementësh ku nyjet janë elementët elektrik dhe lidhëzat janë përcjellësat.

Në qëndër të LMD ështe supozimi se: **Dimensionet fizike të komponentëve të qarkut janë shumë më të vogla se gjatësia e valës elektromagnetike të sinjaleve që transmetohen në qark**. 

Kjo na lejon të bëjmë disa thjeshtime thelbësore:
1. **Neglizhimi i përhapjes së fushave elektromagnetike** – Fushat elektrike dhe magnetike **qëndrojnë brenda përbërësve** (si rezistenca, kondensatorë, dhe bobina).
2. **Rrymat dhe tensionet përkufizohen si vlera pikësore** – Në vend që të analizojmë vektorët e fushave në secilën pikë të hapësirës, ne përdorim vlera skalare si rryma dhe tensioni për çdo komponent.
3. **Përdorimi i elementeve të idealizuar**:
   - **Rezistenca** modelon disipimin e energjisë për shkak të rezistencës së materialit.
   - **Kondensatori** modelon ruajtjen e ngarkesës elektrike.
   - **Bobina** modelon ruajtjen e energjisë në fushën magnetike.
   - **Burim Voltazhi** modelon një burim energjie elektrike që ruan tensionin konstant mes dy terminaleve.
   - **Burim Rryme** modelon një burim energjie elektrike që ruan rrymën pergjatë tij konstante.


### **Nga Elektromagnetizmi në Elektronikë**
Duke përdorur **LMD**, arrijmë tek **analiza e qarqeve elektrike**, ku përdorim ligjet e **Kirchhoff-it për tensionin dhe rrymën** në vend të ekuacioneve të Maxwell-it:


<div style="border:6px solid #007acc; background-color:#ddd; color:#000; border-radius: 5px; padding: 10px;">

1. **Ligji i Kirchhoff-it për Rrymën (KCL)** – *Shuma algjebrike e të gjitha rrymave që hyjnë në një nyje është e barabartë me shumën e të gjitha rrymave që dalin prej saj.* Matematikisht formulohet si:

   $$ \sum I_{\rm in} = \sum I_{\rm out}$$

Ky ligj bazohet në ruajtjen e ngarkesës, që do të thotë se asnjë ngarkesë elektrike nuk mund të krijohet apo të shkatërrohet në një nyje. Një nyje është një pikë në qark ku takohen dy ose më shumë përçues. Biem dakort që rrymat hyrëse në nyje janë pozitive (bien ngarkesë elektrike) ndërsa ato dalëse janë negative (largojnë ngarkesë elektrike).

</div>

<div style="border:6px solid #007acc; background-color:#ccc; color:#000; border-radius: 5px; padding: 10px;">

2. **Ligji i Kirchhoff-it për Tensionin (KVL)**. *Shuma algjebrike e të gjitha tensioneve në një lak të mbyllur të një qarku elektrik është gjithmonë zero.* Matematikisht formulohet si:

  $$ \sum U_i = 0 $$

Ky ligji bazohet në ruajtjen e energjisë, që do të thotë se tensioni i dhënë nga burimet duhet të shpërndahet plotësisht nëpër elementet e tjera në qark. Kjo nënkupton që kur një ngarkesë elektrike lëviz rreth një rruge të mbyllur në qark, ajo nuk mund të grumbullojë apo humbasë energji përveçse nga elementet e qarkut (si rezistencat, kondensatorët, etj.).

</div>


Këto ligje, së bashku me modelet e elementeve të përqendruara, na mundësojnë **analizën e qarqeve elektrike** duke përdorur **superpozimin, ekuivalentin e Thevenin-it dhe Norton-it**, si dhe metoda të tjera që do të trajtojmë.

#### **Përshkrimi Elektromagnetik (Ekuacionet e Maxwell-it)**  
Forca elektromotorre e baterisë krijon një **diferencë potenciale $U$** në skajet e telit përcjellës inkandeshent, dhe si rrjedhojë krijohet një **fushë elektrike $\mathbf{E}$** përgjatë tij. Sipas ekuacioneve të Maxwell-it:

$$ \nabla \times \mathbf{E} = -\frac{\partial \mathbf{B}}{\partial t} $$

Ky ekuacion tregon se **ndryshimi i fushës magnetike** krijon një **fushë elektrike**, por në rastin e një **rryme DC të vazhduar**, fusha magnetike është konstante dhe ky term bëhet zero.  

Fusha elektrike $\mathbf{E}$ në përçues shkakton **lëvizjen e elektroneve**, duke krijuar një **rrymë elektrike $I$)** që përshkon përçuesin. Nga **ekuacioni i vazhdueshmërisë**:

$$ \nabla \cdot \mathbf{J} = -\frac{\partial \rho}{\partial t} $$

Meqë **rryma është konstante në kohë**, kemi që:

$$ \nabla \cdot \mathbf{J} = 0 $$

Kjo do të thotë që **rryma që hyn në një seksion të përçuesit është e barabartë me rrymën që del prej tij** – ky është thelbi i **Ligjit të Kirchhoff-it për rrymën (KCL)** në elektronikë.

Nga ana tjetër, fushat elektrike krijojnë një **diferencë potenciali $U$** përgjatë përçuesit. Sipas **ligjit të Ohm-it në formën diferenciale**:

$$ \mathbf{E} = \rho \mathbf{J} $$

ku **$\rho$** është rezistenca specifike e materialit. Integrimi i këtij ekuacioni përgjatë gjatësisë së përçuesit $\ell$ jep:

$$ U = I\, R $$

kjo është pikërisht **Ligji i Ohm-it**, që përdoret në analizën e qarqeve elektrike.

#### **Përshkrimi me Modelin e Përqendruar (LMD)**
Në vend që të analizojmë vektorët $\mathbf{E}$ dhe $\mathbf{B}$ në çdo pikë të hapësirës, mund të përdorim modelin e përqendruar ku përçuesi modelohet si një **rezistencë $R$** dhe trajtojmë tensionin dhe rrymën si vlera skalare.

#### **Përfundimi**
- **Ekuacionet e Maxwell-it** përshkruajnë sjelljen e përgjithshme të fushave elektrike dhe magnetike, por janë të vështira për t'u përdorur në analiza të thjeshta.
- **Disiplina LMD** na lejon të modelojmë përçuesit si **rezistenca, kondensatorë dhe bobina** me vlera të pikësore të tensionit dhe rrymës.
- **Ligjet e Kirchhoff-it (KVL dhe KCL)** burojnë nga ekuacionet e Maxwell-it dhe ruajtja e energjisë dhe ngarkesës.

Kjo qasje na lejon të analizojmë qarqet në mënyrë të thjeshtë dhe efikase, duke përdorur metoda të tilla si **Superpozimi, Thevenin dhe Norton**, të cilat do të trajtojmë më poshtë.

# **Elementët Bazë të Një Qarku Elektrik**  

Një **qark elektrik** përbëhet nga **elementë bazë pasivë dhe aktivë** që modelojnë sjelljen e energjisë elektrike. Këto elemente kanë **veti të ndryshme** dhe **modele matematikore** që përdoren në analizën e qarqeve elektrike.  

## **Elementët Pasivë**  

Elementët **pasivë** nuk gjenerojnë energji por e **ruajnë, e shpërndajnë ose e vonojnë atë**. Këta përfshijnë **rezistencat, kondensatorët dhe bobinat**.

### **Rezistenca ($R$)**
- **Çfarë modelon?**  
  - Modelon **humbjen e energjisë** për shkak të rezistencës së përçuesve (p.sh., përçues bakri, tela, rezistencat).
  - Modelon **"fërkimin elektrik"**, ku energjia elektrike shndërrohet në **nxehtësi** (efekti Joule).
  - Simboli: <img src="figs/lec3/R.svg"/>
  
- Varësia $U-I$ është **Ligji i Ohm-it:**  
  $$ U = I\,R $$
  ku: $U$ është tensioni i aplikuar mbi rezistence, $I$ është rryma që kalon nëpër të dhe $R$ është rezistenca në ohm ($\Omega$).

- **Veti të Rezistences:**
  - Në **DC (Rrymë të vazhduar)**: Rezistenca është konstante.
  - Në **AC (Rrymë alternative)**: Nuk ndikon në frekuencën apo fazën e sinjalit.
  - **Fuqia e Çliruar** nga një rezistenca:
    $$ P = U\,I = I^2 \,R = \frac{U^2}{R} $$

### **Kondensatori ($C$)**
- **Çfarë modelon?**  
  - Modelon **ruajtjen e energjisë në fushën elektrike** midis dy pllakave të tij.
  - Përdoret për **filtrim, stabilizim të tensionit, dhe ruajtje të ngarkesës elektrike**.
  - Simboli: <img src="figs/lec3/C.svg"/>

- **Lidhja ndërmjet tensionit dhe rrymës:**  
  $$ I = C \,\frac{dU}{dt}$$
  ku $C$ është kapaciteti në farad ($F$).
  - **Në DC**: Në gjëndje stacionare ($dU/dt=0$), vepron si qark i hapur ku $I=0$ (asnjë rrymë nuk kalon).
  - **Në AC**: Ofron një **reaktancë kapacitive** (inpendancë) që zvogëlohet me frekuencën:
    $$ X_C = \frac{1}{\omega C}, \quad \text{me} \quad \omega = 2\pi f $$
  - **Energjia Elektrike** që ruan kur ka differencë potencial $U$ është:
    $$ E = \frac{C\, U^2}{2} $$
  - Nuk ndryshon frequencën e sinjalit por sfazon rrymën nga tensioni. Për kondensatorin, rryma është përpara në fazë nga tensioni me $\varphi=\frac{\pi}{2}$.


### **Bobina ($L$)**
- **Çfarë modelon?**  
  - Modelon **ruajtjen e energjisë në fushë magnetike**.
  - Përdoret për **filtrim, stabilizim të rrymës, dhe kufizimin e ndryshimeve të papritura në rrymë**.
  - Simboli: <img src="figs/lec3/L.svg"/>

- **Lidhja ndërmjet tensionit dhe rrymës:**  
  $$ U = L \frac{dI}{dt}  $$
  ku $L$ është induktanca në henri ($H$).
  - **Në DC**: Në gjëndje stacionare ($dI/dt=0$) vepron si një përçues i zakonshëm.
  - **Në AC**: Ofron një **reaktancë induktive** që rritet me frekuencën:
    $$ X_L = \omega L $$
  - **Energjia Magnetike** që ruan kur ka kalon rryma $I$ është:
    $$ E = \frac{L\, I^2}{2} $$
  - Nuk ndryshon frequencën e sinjalit por sfazon rrymën nga tensioni. Për bobinën, rryma është në vonesë faze me $\varphi=\frac{\pi}{2}$ nga tensioni.


## **Elementët Aktivë**  

Elementët **aktivë** mund të **gjenerojnë ose amplifikojnë energji elektrike**. Këto përfshijnë burimet e tensionit dhe rrymës, transistorët, dhe amplifikatorët.

### **Burimi i Tensionit**
- **Çfarë modelon?**  
  - Modelon një **element që krijon tension konstant** pavarësisht rrymës.
- **Llojet e burimeve të tensionit:**
  - **Burimi ideal**: Gjithmonë ofron një tension konstant $U$.
  - **Burimi real**: Ka një **rezistencë të brendshme** që ndikon në tensionin dalës.
  - Simboli: <img src="figs/lec3/Us.svg"/> (me romb shënohen burimet që i komandohet tensioni)

### **Burimi i Rrymës**
- **Çfarë modelon?**  
  - Modelon një **element që krijon një rrymë konstante**, pavarësisht tensionit.
- **Llojet e burimeve të rrymës:**
  - **Burimi ideal**: Gjithmonë ofron një rrymë konstante $I$.
  - **Burimi real**: Ka një **rezistencë të brendshme** që ndikon në rrymën dalëse.
  - Simboli: <img src="figs/lec3/Is.svg"/>  (me romb shënohen burimet që i komandohet rryma)


### **Transistorët (BJT & FET)**
- **Çfarë modelojnë?**  
  - Modelojnë **amplifikatorë të sinjalit**, **celës elektronikë**, dhe përdoren në logjikën digjitale.
  - Për shembull, një transistor bipolar **BJT** ka tre terminale: **Baza (B), Kolektori (C), Emiteri (E)**.
  - Një transistor me efekt fushe **FET** ka **Portën apo Gate (G), mbledhësin apo drain (D), dhe burimin apo Source (S)**.
  - Simboli: <img src="figs/lec3/BJT_FET.svg"/>

## **Çfarë është një Qark Real (Fizik)?**

Një qark **real ose fizik** përbëhet nga pajisje dhe elemente **konkrete** të lidhura fizikisht me njëra-tjetrën, si:

- Burime (bateri, gjeneratorë)
- Tela, rezistenca, kondensatorë, bobina, tranzistorë realë
- Instrumente matëse si voltmetra, ampermetra
- Kabllo lidhëse, ndërprerës fizikë, pllaka prototipizuese (breadboards), PCB, etj.

Ky qark ka **dimensione fizike** dhe është subjekt i ligjeve të fizikës. Për rrjedhojë, pershkrimi i tyre me elementë ideal mund te mos jete i mjaftueshem dhe nevojitet përfshirja e parametrave **parazitare** si dhe efekte të tjera që nuk merren parasysh nga elementët ideale.

### **Karakteristikat e një qarku real:**

- **Rezistencë parazite**: Çdo tel ka rezistencë (të vogël por jo zero).
- **Kapacitetet dhe induktancat parazite**: Çdo tel dhe çdo komponent ka induktivitet dhe kapacitet parazit (sado i vogël por jo zero).
- **Humbjet e Energjisë**: Nxehtësia, rrezatimi elektromagnetik, rezistenca e përçuesve, oksidimi i kontakteve.
- **Efektet Termike**: Rezistenca ndryshon me temperaturën.
- **Sinjalet reale**: Burimet nuk janë ideale; tensioni dhe rryma mund të bien në varësi të ngarkesës.

Eshtë e qartë nga figura më lart që psh. telat lidhës që janë pranë njëri-tjetrit formojnë një strukturë ku përcjellësat ndahen mes tyre nga një dielektrik (qoftë ajri ose izolatori i telave). Kjo është struktura tipike e një kondensatori. Pra qarku fizik, parvarësisht se mund të mos ketë kondensator të lidhur në të, vetëm natyra se si lidhen komponentet (telat dhe elementët e tjerë të qarkut) në mënyrë të pashmangshme prodhon kapacitet parazit (pra të pa dëshiruar). Për më tepër, telat dhe komponentet e tjera formojnë qark të mbyllur. Këto formojnë në fakt spirat e një bobine. Rrjedhimisht, qarku real prodhon një farë induktiviteti parazit, thjeshtë nga natyra e lidhjeve fizike. Nëse këto lidhje janë si në figurën e dytë djathtas, ato nëpërmjet induksionit elektromagnetik mund të interferojnë me njëra tjetrën (si dy bonina të çiftuara). Gjithashtu, nga ana tjetër, kontaktet e telave me breadboard nuk janë perfekte dhe shtojnë një rezistencë ekstra nga ajo të cilën ne modelojmë. 

Të gjitha këto efekte parazitare mundet të jenë të vogla në pvrgjithesi. Megjithatë, siç kemi parë nga transformimet Fourier, sinjalet elektrike periodike që mund të mos jenë me frekuenca të larta, dekompozohen në një spektër që përmban frekuenca të larta, kujtoni psh. sinjalin drejtkëndore. Harmonikat me frekuenca të larta të sinjalit bashkëveprojnë ndryshe me induktivitetin ($X_L=\omega\,L$) dhe kapacitetin ($X_C=1/\omega C$) parazit të qarkut, duke bërë që të përhapen ndryshe. Si rrjedhojë, në varësi se sa i madh është efekti, sinjali i plotë që do të përhapej në këtë qark do të ishte i deformuar. Kjo ndryshon analizen e qarkut.

## **Çfarë është një Skemë Elektrike?**

Një **skemë** është një **model grafik i idealizuar** që përdoret për analizën teorike dhe simulimin e qarqeve elektrike.

Në një skemë, elementët janë **idealizuar plotësisht**, duke mos përfshirë **detajet fizike reale**.

### **Karakteristikat e Modelit Skematik**:

- **Elementë idealë**: Rezistencat, kondensatorët dhe bobinat janë ideale dhe nuk kanë elementë parazitë.
- **Dimensione Jo-Fizike**: Gjithçka është e përqendruar në pika dhe linja pa dimension fizik.
- **Lidhje perfekte**: Tela ideale pa rezistencë dhe kapacitet parazit.
- **Burime ideale**: Burime që japin rrymë dhe tension të pandryshuar pavarësisht nga ngarkesa.
- **Injorimi i kushteve mjedisore**: Nuk merren parasysh efektet e temperaturës, lagështisë, apo zhurmës.

##### **Skemat Elektronike janë Topologjikisht Ekuivalente**

Qarqet elektronike formojnë një strukturë grafi. Një graf është një bashkësi nyjesh (nodes, në këtë rast elementësh elektronikë si rezistenca, kondensatorë, bateri etj...) dhe lidhëzash (edges, si përcjellësat, telat, etj...).

**Skema elektronike mund të duken shumë të ndryshme nga mënyra se si janë vizatuar, megjithatë nësë ato kanë të njëjtin numër nyjesh (elementësh) dhe të njëjtat lidhje mes tyre, pra nëse janë i njëjti graf, këto skema janë ekuivalente. Pra në na lejohet ta ndryshojmë vizatimin e një skeme elektronike si të duam, për aq kohë sa nuk ndryshojmë lidhjet dhe elementet. Themi se skemat janë topologjikisht ekuivalente.**

Me pak fjalë, skemat elektronike nuk varet nga mënyra se si vendosen elementet në hapësirë apo nga gjatësia dhe forma e lidhëzave.
</div>

# Lidhjet në Seri dhe në Paralel

##### **Lidhja në Seri**

Në lidhjen në seri, komponentët lidhen **njëri pas tjetrit**, në të njëjtën linjë elektrike:

- **Rryma** $I$ është e njëjtë në çdo komponent.
- **Tensioni** ndahet mes komponentëve.

##### **Lidhja në Paralel**

Në lidhjen paralele, komponentët lidhen në dy pika të përbashkëta, duke krijuar **degëzime**:

- **Tensioni** është i njëjtë në çdo komponent.
- **Rryma** ndahet mes komponentëve.

| Komponenti | Lidhja në Seri | Lidhja në Paralel |
|------------|----------------|-------------------|
| **Rezistencë (R)** | $R_t=R_1 + R_2 + \dots + R_n$ | $\frac{1}{R_t} = \frac{1}{R_1} + \dots + \frac{1}{R_n}$ |
| **Kondensator (C)** | $\frac{1}{C_t} = \frac{1}{C_1} + \dots + \frac{1}{C_n}$ | $C_t=C_1 + C_2 + \dots + C_n$ |
| **Bobinë (L)** | $L_t=L_1 + L_2 + \dots + L_n$ | $\frac{1}{L_t} = \frac{1}{L_1} + \dots + \frac{1}{L_n} $ |
| **Burim tensioni (V)** | $U_t=U_1 + U_2 + \dots + U_n$ | Duhet të kenë të njëjtin tension |
| **Burim rryme (I)** | Duhet të kenë të njëjtën rrymë | $I_t=I_1 + I_2 + \dots + I_n$ |


Këto formula dhe parime janë **themelore** për thjeshtimin dhe analizën e qarqeve elektrike.

# Analiza e Qarqeve Elektrike

Qëllimi i **analizës së qarqeve elektrike** është të përcaktojë në mënyrë të saktë **sjelljen elektrike të qarqeve**, duke llogaritur tensionet, rrymat dhe fuqinë e shpërndarë në komponentët e ndryshëm. 

Analiza e qarqeve është mjeti kryesor për të **kuptuar, projektuar dhe kontrolluar** sjelljen e çdo qarku elektrik dhe elektronik, duke siguruar që qarqet të funksionojnë në mënyrë të saktë, të sigurt dhe efikase.


## **Qëllimet Kryesore të Analizës së Qarqeve:**

1. **Parashikimi i sjelljes së qarkut**
   - Identifikimi i saktë i tensioneve dhe rrymave në çdo pikë.
   - Kontrollimi nëse qarku punon sipas pritshmërive teorike.

2. **Projektimi dhe optimizimi i qarqeve**
   - Zgjedhja e komponentëve me vlera optimale.
   - Optimizimi i konsumit dhe efikasitetit energjetik.

3. **Identifikimi i problemeve dhe parandalimi i dëmtimeve**
   - Identifikimi i kushteve ekstreme (rrymë, tension).
   - Parandalimi i mbingarkesave dhe dëmtimeve të komponentëve.

4. **Verifikimi teorik para implementimit fizik**
   - Analiza teorike kursen kohë, para, dhe mundin e korrigjimit të gabimeve në faza të mëvonshme.

5. **Diagnoza e problemeve në qarqe ekzistuese**
   - Analiza ndihmon në identifikimin dhe zgjidhjen e problemeve në qarqet reale.


###**Metodat kryesore të analizës:**

- Ligjet e Kirchhoff-it (**KVL, KCL**)
- Analiza e nyjeve (**Node Analysis**)
- Analiza e superpozimit (**Superposition**)
- Metodat ekuivalente (**Thevenin, Norton**)


## **Metoda 1: Ligjet e Kirchhoff-it**

Algoritmi i Përgjithshëm i Zgjidhjes me KVL & KCL:

1. Identifiko nyjet dhe laket e qarkut.
2. Vendos drejtim arbitrar për rrymat dhe polaritet për tensionet.
3. Apliko KCL për çdo nyje (siguron lidhjen e rrymave).
4. Apliko KVL për çdo lak të pavarur (siguron lidhjen e tensioneve).
5. Apliko varësinë U-I (volt-amper) për çdo element në qark.
6. Zgjidh sistemin e ekuacioneve.

## **Metoda 2: Kombinimi i elementëve (seri dhe paralel)**

Kjo metodë konsiston në **thjeshtimin e qarkut**, duke kombinuar gradualisht elementët (rezistenca, kondensatorë, bobina) të lidhura në **seri** apo **paralel**, derisa të arrihet në një qark më të thjeshtë që mund të analizohet lehtë.

### **Procedura hap pas hapi:**

#### **Hapi 1: Identifikimi i Elementëve të Kombinueshëm**
- Zgjidh qarqet ose segmentet ku komponentët janë të lidhur në mënyrë të qartë në **seri** ose **paralel**.

#### **Hapi 2: Kombinimi i Elementëve sipas Rregullave**
Apliko rregullat e mëposhtme sipas llojit të elementëve:

- **Rezistenca:**
  - Seri: $R_{eq} = R_1 + R_2 + \dots + R_n$
  - Paralel: $\frac{1}{R_{eq}} = \frac{1}{R_1} + \frac{1}{R_2} + \dots + \frac{1}{R_n}$

- **Kondensatorë:**
  - Seri: $\frac{1}{C_{eq}} = \frac{1}{C_1} + \frac{1}{C_2} + \dots + \frac{1}{C_n}$
  - Paralel: $C_{eq} = C_1 + C_2 + \dots + C_n$

- **Bobina:**
  - Seri: $L_{eq} = L_1 + L_2 + \dots + L_n$
  - Paralel: $\frac{1}{L_{eq}} = \frac{1}{L_1} + \frac{1}{L_2} + \dots + \frac{1}{L_n}$

#### **Hapi 3: Thjeshtimi Progresiv i Qarkut**
- Zëvendëso elementët e kombinuar me elementin ekuivalent.
- Përsërite kombinimin hap pas hapi, duke e bërë qarkun gjithnjë e më të thjeshtë.

#### **Hapi 4: Analiza e Qarkut të Thjeshtuar**
- Analizo qarkun e thjeshtuar për të gjetur madhësitë e dëshiruara (rryma, tensione).

#### **Hapi 5: Rikthimi në Qarkun Origjinal**
- Kthehu prapa duke përdorur madhësitë e gjetura në hapin 4, për të gjetur madhësitë në çdo komponent origjinal.

## **Metoda 3: Analiza e Nyjeve (Node Analysis)**

Analiza e nyjeve është një aplikim konkret dhe sistematik i ligjit të Kirchhoff-it për rrymat (KCL), që bën të mundur analizën e qarqeve nëpërmjet përcaktimit të tensioneve në nyjet e qarkut.

Qëllimi kryesor është të përcaktohen **tensionet e nyjeve**, nga të cilat llogariten lehtësisht rrymat në çdo element.


### **Procedura hap pas hapi:**

#### **Hapi 1: Identifiko Nyjen Referencë (Ground)**  
- Zgjidh një nyje si **pikë referimi** (masë ose ground). Zakonisht është pika më e poshtme ose ajo me më shumë lidhje. Simboli i ground është <img src="figs/lec3/GND.svg"/> dhe shënohet edhe me GND.

#### **Hapi 2: Etiketo Tensionet e Nyjeve të Tjera**  
- Jepu emër nyjeve të tjera me tensione të panjohura: $V_1, V_2, \dots, V_n $.

#### **Hapi 3: Shkruaj KCL në Çdo Nyje**  
- Për secilën nyje apliko KCL:
$$ \sum I_{\text{hyrëse}} = \sum I_{\text{dalëse}} $$

- **Shprehim rrymat si funksion i tensioneve të nyjeve?**  
Nëse kemi rezistencë midis dy nyjeve me $V_a$ dhe $V_b$, rryma është:
$$ I = \frac{V_a - V_b}{R} $$

#### **Hapi 4: Zgjidh Sistemin e Ekuacioneve**  
- Nga hapi 3 del një sistem ekuacionesh lineare me tensionet e panjohura.
- Zgjidh këto ekuacione me metodat standarde (algjebra lineare).

#### **Hapi 5: Llogarit Rrymat në Degë**  
- Pasi gjen tensionet e nyjeve (psh $V_a$ dhe $V_b$), përdori ato për të llogaritur rrymat e secilit element:
$$ I_{R} = \frac{V_a - V_b}{R} $$

### **Kur përdoret Metoda e Analizës së Nyjeve?**

- Veçanërisht efikase kur ka **pak nyje** dhe shumë degë.
- Shumë e përshtatshme për qarqe lineare komplekse me disa burime dhe elemente pasivë.
- Është metodë shumë intuitive dhe e përshtatshme për analizën kompjuterike (llogaritje numerike).


#### Shembull 1

<img src="figs/lec3/circ1_node.svg" align="center" width=400 style="margin-left:auto; margin-right:auto"/>

$$
KCL = \begin{cases} 
b: & \frac{U_0-V_b}{R_1}=\frac{V_b}{R_2} + \frac{V_b - V_d}{R_3} \\
d: & \frac{V_b-V_d}{R_3} + \frac{U_0-V_d}{R_4}= \frac{V_d}{R_5}
\end{cases}
$$

$$
\begin{cases} 
b: & \left[\frac{1}{R_1}+\frac{1}{R_2}+\frac{1}{R_3} \right]\, V_b - \left[\frac{1}{R_3}\right]\,V_d =\frac{U_0}{R_1} \\
d: & -\left[\frac{1}{R_3}\right]\,V_b + \left[\frac{1}{R_3}+\frac{1}{R_4}+\frac{1}{R_5} \right]\, V_d = \frac{U_0}{R_4}
\end{cases}
$$

$$
\begin{cases} 
\;\; g_1\, V_b - g_2\,V_d =\frac{U_0}{R_1} \\
-g_2\,V_b + g_1\, V_d = \frac{U_0}{R_4}
\end{cases}
$$

$$ G\,V = S $$

## **Metoda 4: Metoda e Superpozimeve**

Nga shembujt që pamë më lart, vërejmë se potenciali në çdo nyje $V_i$ jipet si kombinim linear i te gjithë burimeve qofshin tensioni apo rryme. Me pak fjalë:

$$ V_i= \sum_k \alpha_k\, U_k + \sum_l \beta_l\, I_l$$

ku $U_k$ dhe $I_l$ janë burimet e tensioneve dhe rrymave, respektivisht.

> Metoda Superpozimeve supozon burime të pavarura (tensioni ose rryme), si rrjedhojë output i qarkut përcaktohet duke mbledhur përgjigjet e secilit burim sikur secili të punonte i vetëm.

Kjo metodë është e lidhur ngushtë me qarkun linearë.

#### **Çfarë është një Qark Linear?**

 
<div style="border:6px solid #f00; background-color:#ddd; color:#000; border-radius: 5px; padding: 10px;">

##### **Linearitetit te një Funksioni (Kujtesë)**

$$
\text{Lineariteti} \Rightarrow \begin{cases} \text{Homogjenitet} \\ \text{Superpozim} \end{cases}
$$

Supozojmë një funksion i disa variablave $f(x_1, x_2, ..., x_n)$.

- **Vetia e Homogjenitet:**
 $$f(\alpha\, x_1, \alpha\,x_2, ..., \alpha\,x_n) = \alpha\,f(x_1, x_2, ..., x_n) \qquad \left[f(\alpha\, x) = \alpha\, f(x) \right] $$

- **Vetia e Superpozim:** 
 $$f(x_1+y_1, x_2+y_2, ..., x_n+y_n) = f(x_1, x_2, ..., x_n) + f(y_1, y_2, ..., y_n) \qquad \left[ f(x+y) = f(x) + f(y) \right]$$


</div>

<br>

Një qark quhet **linear** nëse marrëdhënia midis tensioneve dhe rrymave në të është lineare. Qarku linear ka karakteristika të thjeshta:

- Superpozimi vlen plotësisht.
- Tensionet dhe rrymat e tij mund të përshkruhen nga ekuacione lineare.

> **Vëreni që ekuacionet që jane shfaqur ne ushtrimet qe kemi zgjidhur janë lineare!**

## **Metoda e Ekuivalencës**

Metoda e ekuivalencës konsiston në zëvendësimin e një pjese të një qarku elektrik kompleks me një qark më të thjeshtë, por plotësisht **ekuivalent në aspektin e sjelljes elektrike**.

<img src="figs/lec3/equiv1.png" align="center" width=700 style="margin-left:auto; margin-right:auto"/>

*Ref: Anant Agarwal and Jeffrey Lang, course materials for 6.002 Circuits and Electronics, Spring 2007. MIT OpenCourseWare (http://ocw.mit.edu/).*


Dy metodat më të njohura të ekuivalencës janë:

- **Teorema e Thevenin-it** (Burim tensioni + rezistencë)
- **Teorema e Norton-it** (Burim rryme + rezistencë)

<br>

<div style="border:6px solid #f00; background-color:#ddd; color:#000; border-radius: 5px; padding: 10px;">

#### **Teorema e Thevenin-it**
 
Çdo qark linear që përmban burime dhe rezistenca mund të zëvendësohet nga një **burim të vetëm tensioni $V_{TH}$** në seri me një **rezistencë të vetëm $R_{TH}$**, duke parë nga terminalet e qarkut.

<img src="figs/lec3/Thevenin.svg" align="center" width=200 style="margin-left:auto; margin-right:auto"/>

##### **Parametrat e Thevenin-it:**

- **Tensioni $V_{TH}$**: Tensioni në terminalet e qarkut të hapur (open-circuit voltage).
- **Rezistenca $R_{TH}$**: Rezistenca ekuivalente në terminalet e qarkut duke shuar burimet:

<img src="figs/lec3/thevenin.png" align="center" width=600 style="margin-left:auto; margin-right:auto"/> 

</div>

<br>

<div style="border:6px solid #f00; background-color:#ddd; color:#000; border-radius: 5px; padding: 10px;">

#### **Teorema e Norton-it**
 
Çdo qark linear mund të zëvendësohet me një **burim të vetëm rryme $I_{N}$** në paralel me një **rezistencë të vetme $R_{N}$**, duke parë nga terminalet e qarkut.

<img src="figs/lec3/Norton.svg" align="center" width=200 style="margin-left:auto; margin-right:auto"/>

##### **Parametrat e Norton-it:**

- **Rryma $I_{N}$**: Rryma në terminalet e qarkut të shkurtër (short-circuit current).
- **Rezistenca $R_{N}$**: Ka të njëjtën vlerë si rezistenca e Thevenin-it ($R_N=R_{TH}$).

<img src="figs/lec3/norton.png" align="center" width=600 style="margin-left:auto; margin-right:auto"/> 

</div>

<img src="figs/lec3/thevenin3.png" align="center" width=500 style="margin-left:auto; margin-right:auto"/> <img src="figs/lec3/thevenin4.png" align="center" width=500 style="margin-left:auto; margin-right:auto"/> 



#### **Konvertimi mes Thevenin-it dhe Norton-it**

Dy ekuivalentët mund të kalojnë lehtë nga njëri në tjetrin:

- Nga Thevenin në Norton:
$$ I_N=\frac{V_{TH}}{R_{TH}}, \quad R_N=R_{TH} $$

- Nga Norton në Thevenin:
$$ V_{TH}=I_N \cdot R_N, \quad R_{TH}=R_N $$


### **Algoritmi për aplikimin e Teoremave të Ekuivalencës**

1. **Identifikimi i segmentit të qarkut për zëvendësim**
   - Zgjidhni një pjesë të qarkut për ta zëvendësuar me ekuivalentin Thevenin ose Norton.

2. **Gjeni $V_{TH}$ ose $I_{N}$**
   - $V_{TH}$: Matni tensionin në terminalet e hapura të segmentit.
   - $I_{N}$: Matni rrymën në terminalet e shkurtuara.

3. **Gjeni rezistencën $R_{TH}=R_{N}$**  
   - Shuani burimet (burimet e tensionit bëhen qark i shkurtër, burimet e rrymës bëhen qark i hapur).
   - Matni rezistencën ekuivalente në terminalet e segmentit.

4. **Zëvendëso segmentin origjinal me ekuivalentin**
   - Përdor ekuivalentin (Thevenin ose Norton) për analiza më të thjeshta të qarkut.


### **Lidhja me Metodën e Superpozimit**

Në qarqet lineare, teoremat e Thevenin dhe Norton funksionojnë shkëlqyeshëm me metodën e superpozimit. Superpozimi thotë:

> Në një qark linear me disa burime, efekti total në çdo pikë është shuma algjebrike e efekteve të çdo burimi individual, duke shuar burimet e tjera.

### **Algoritmi i kombinuar (Superpozim + Thevenin/Norton):**

- **Hapi 1**: Analizoni efektin e çdo burimi individualisht (shuani burimet e tjera).
- **Hapi 2**: Për çdo analizë individuale, përdorni ekuivalentin Thevenin ose Norton për thjeshtim.
- **Hapi 3**: Llogaritni rrymat dhe tensionet në secilin rast.
- **Hapi 4**: Mblidhni rezultatet (superpozim) për të gjetur zgjidhjen përfundimtare.


### **Përmbledhje e shkurtër:**

- **Qark linear**: Mund të zëvendësohet nga ekuivalentë të thjeshtë (Thevenin/Norton).
- **Thevenin**: Burim tensioni + rezistencë në seri.
- **Norton**: Burim rryme + rezistencë në paralel.
- **Superpozimi**: Ndërthuret shkëlqyeshëm me metodat e ekuivalencës për analiza të thjeshta dhe efektive.

## **Ushtrime Experimentale**

## **Ushtrimi 1: Matja dhe analiza e rezistencës së vërtetë**

### **Qëllimi:**
- Të demonstrohet efekti i **rezistencës parazite** në telat dhe komponentët fizikë.

### **Pajisjet laboratorike:**
- Multimetër digjital (ose analog)
- Rezistenca nominale (p.sh. 10 Ω, 100 Ω, 1 kΩ)
- Tela me gjatësi të ndryshme dhe breadboard.

### **Procedura:**
- Matni rezistencën nominale në laborator dhe krahasojeni atë me vlerën e përcaktuar nga ngjyra.
- Matni rezistencën e telave me gjatësi të ndryshme (p.sh., 10 cm, 50 cm, 1 m).

### **Analiza dhe raportimi:**
- Krahaso rezistencën e matur me atë teorike dhe diskuto efektet e rezistencës parazite.

<br>

## **Ushtrimi 2: Efekti i Kapacitetit Parazit në qarqe me frekuenca të larta**

### **Qëllimi:**
- Të demonstrohet efekti i kapacitetit parazit në sjelljen e rezistencave dhe qarqeve në frekuencë të lartë.

### **Pajisjet laboratorike:**
- Osciloskop
- Gjenerator funksionesh (për frekuenca > 1 MHz)
- Rezistenca të mëdha (p.sh., 100 kΩ, 1 MΩ)
- Breadboard dhe tela lidhës

### **Procedura:**
- Aplikoni sinjale sinusoidale me frekuenca në intervalin 1 kHz – 10 MHz në rezistencën e madhe.
- Matni tensionin dhe fazën mes tensionit dhe rrymës për rezistenca të ndryshme dhe frekuenca të ndryshme.

### **Analiza dhe raportimi:**
- Krahaso rezultatet eksperimentale me parashikimin teorik duke përfshirë kapacitetin parazit (~0.2–1 pF).
- Diskutoni pse efekti bëhet i dukshëm në frekuenca të larta.

<br>

## **Ushtrimi 3: Induktanca Parazite dhe Impedanca në Qarqe RLC**

### **Qëllimi:**
- Të shfaqet efekti i induktancës parazite në një rezistencë të zakonshme aksiale në qarqe me frekuencë të lartë.

### **Pajisjet laboratorike:**
- Osciloskop
- Gjenerator funksionesh (0.1–10 MHz)
- Rezistenca standarde (p.sh., 100 Ω–1 kΩ)
- Kondensatorë standard (p.sh., 10 nF, 100 nF)
- Breadboard dhe tel lidhës

### **Procedura:**
- Ndërtoni një qark të thjeshtë RC dhe aplikoni sinjale me frekuencë të lartë.
- Matni tensionin dhe fazën në rezistencë dhe kondensator.
- Përsëritni matjet për rezistenca dhe kondensatorë të ndryshëm.

### **Analiza dhe raportimi:**
- Diskuto diferencat e matjeve nga modeli ideal teorik RC, duke përfshirë induktancën parazite (~10–20 nH).

<br>

## **Ushtrimi 4: Krahasimi eksperiment-teori për Superpozimin**

### **Qëllimi:**
- Të demonstrohet vërtetësia praktike e metodës së superpozimit në qark real.

### **Pajisjet laboratorike:**
- Multimetër dhe burime të pavarura (bateri ose burime DC laboratorike)
- Disa rezistenca (100 Ω, 220 Ω, 470 Ω)
- Breadboard dhe tel lidhës

### **Procedura:**
- Ndërtoni një qark me dy burime të ndryshme tensioni dhe disa rezistenca.
- Matni tensionin dhe rrymën në degë duke aktivizuar vetëm një burim dhe duke fikur tjetrin.
- Përsëriteni këtë matje për secilin burim.

### **Analiza dhe raportimi:**
- Krahaso rezultatin e matjeve me rezultatin teorik të llogaritur nga superpozimi.
- Diskutoni saktësinë dhe burimet e mundshme të gabimeve eksperimentale.

<br>

## **Ushtrimi 5: Ekuivalenca Thevenin dhe Norton në Praktikë**

### **Qëllimi:**
- Të shfaqet ekuivalenca praktike e qarqeve Thevenin dhe Norton me qarqet reale.

### **Pajisjet laboratorike:**
- Burime tensioni dhe rryme (ose burim tensioni me rezistencë seri si burim rryme)
- Rezistenca standarde (220 Ω, 1 kΩ, 2.2 kΩ)
- Multimetër
- Breadboard dhe tel lidhës

### **Procedura:**
- Ndërtoni një qark të thjeshtë dhe llogaritni ekuivalentin e tij Thevenin dhe Norton teorikisht.
- Ndërtoni qarkun ekuivalent praktikisht dhe matni tensionin dhe rrymën e ngarkesës për të dy rastet.
- Krahasojini këto rezultate.

### **Analiza dhe raportimi:**
- Diskutoni pse matjet praktike përputhen me ekuivalentët teorikë.
- Çfarë efektesh parazite (rezistenca e telave, lidhjet jo ideale) ndikojnë në rezultatet?
