```python
rows = 4
cols = 4
with schemdraw.Drawing() as d:
    d.config(inches_per_unit=0.3)
    u = d.unit
    for i in range(rows):
        for j in range(cols):
            elm.Resistor().at((i*u,j*u))
    
    for i in range(1,rows):
        for j in range(-1,cols):
            elm.Resistor().at((i*u,j*u)).up()
    x_min, x_max = 0, rows*u
    y_min, y_max = -u, cols*u
    elm.Line().at((x_min - u, y_min - u)).right(x_max - x_min + 2*u)
    elm.Line().up(y_max - y_min + 2 * u)
    elm.Line().left(x_max - x_min + 2 * u)
    elm.Line().down(y_max - y_min + 2 * u)
    # Add DC voltage source (bottom left)
    V_dc = elm.SourceV().at((x_min - 2 * u, y_min + u)).right().label("V (DC)")
    elm.Line().right().to((x_min - u, y_min + u))  # Connect to square

    # Add AC voltage source (top left)
    V_ac = elm.SourceSin().at((x_min - 2 * u, y_max - u)).right().label("V (AC)")
    elm.Line().right().to((x_min - u, y_max - u))  # Connect to square
```

```python
import schemdraw
import schemdraw.elements as elm
elm.style(elm.STYLE_IEC)

L = 2
with schemdraw.Drawing() as d:
    Ib = elm.SourceI().label(r'$I_b$').length(L)
    elm.
    elm.Line().at(Ib.end).right().length(L)
    Q  = elm.BjtNpn(circle=True).anchor('base')
    Rc = elm.Resistor().at(Q.collector).up().label(r'$R_c$').length(L)
    elm.Vdd().label('Vcc')
    elm.Line().at(Q.emitter).toy(Ib.start)
    elm.Ground()
    elm.Line().tox(Ib.start)

    d.draw()
```

```python
with schemdraw.Drawing() as d:
    C1 = elm.Capacitor().label(r'$C_1$').length(1.6).idot(open=True)
    #elm.CurrentLabelInline(direction='in')
    R1 = elm.Resistor().up().label(r'$R_1$').idot()
    R2 = elm.Resistor().at(R1.start).down().label(r'$R_2$')
    elm.Line().at(R1.start).right().length(1.5)
    Q  = elm.BjtNpn(circle=True).anchor('base')
    Rc = elm.Resistor().at(Q.collector).toy(R1.end).dot().label(r'$R_c$', loc='bottom')
    elm.Line().tox(R1.end)
    Re = elm.Resistor().at(Q.emitter).toy(R2.end).label(r'$R_e$')
    elm.Ground()
    elm.Line().tox(R2.end)
    elm.Vdd().at(Rc.end).label('Vcc=12V')
    C2 = elm.Capacitor().at(Q.collector).right().label(r'$C_2$').length(1.6).idot().dot(open=True)
    elm.Line().at(Re.start).right().length(0.75).idot()
    Ce = elm.Capacitor().down().toy(Re.end).label(r'$C_e$', loc='bottom')
    elm.Line().tox(Re.end).dot()
    d.save('figs/lec7/CE_amp.svg',transparent=False)
```

```python
with schemdraw.Drawing() as d:
    bjt=elm.BjtNpn(circle=True).label('npn',loc='right').label('baza (B)',loc='base').label('kolektori (C)',loc='collector').label('emiteri (E)',loc='emitter')
    d.move_from(bjt.collector, dx=1.2*d.unit, dy=-0.25*d.unit)
    elm.BjtPnp(circle=True).flip().label('pnp',loc='right').label('baza (B)',loc='base').label('kolektori (C)',loc='collector').label('emiteri (E)',loc='emitter')
    d.save('figs/lec7/BJT_pnp_npn.svg',transparent=False)
```
