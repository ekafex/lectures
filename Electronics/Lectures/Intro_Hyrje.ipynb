# Elektronika (Analoge & Digjitale)

### Objektivi
Kjo lëndë synon t’ju japë **kompetencë funksionale në elektronikë**, në mënyrë që të jeni në gjendje të kuptoni, analizoni dhe përdorni qarqe analoge dhe digjitale në kontekste laboratorike dhe eksperimentale.

Fokusi nuk është thellësia e specializuar, por **kuptimi i qartë i parimeve themelore** dhe aftësia për t’i zbatuar ato praktikisht.


#### Çfarë do bëjm
<div style="border:6px solid #007acc; background-color:#ddd; color:#000; border-radius: 5px; padding: 10px;">

* Analizë qarku
* Sjellje në kohë dhe frekuencë
* Pajisjet gjysmëpërçuese
* Përforcim dhe feedback
* Elektronikë digjitale


Në laborator do të jeni në gjendje të:
* Përdorni oshiloskopin, gjeneratorin e funksioneve dhe burimet e ndryshme te energjisë.
* Realizoni matje dhe t’i krahasoni me simulimet.
* Ndiqni ciklin metodologjik:

> Model → Simulim → Matje → Interpretim

</div>

# Electronics (Analog & Digital)

### Objective
This course aims to provide you with **functional competence in electronics**, enabling you to understand, analyze, and use analog and digital circuits in laboratory and experimental contexts.

The focus is not on specialized depth, but on a **clear understanding of fundamental principles** and the ability to apply them practically.


#### What We Will Cover

<div style="border:6px solid #007acc; background-color:#ddd; color:#000; border-radius: 5px; padding: 10px;">

* Circuit analysis
* Time and frequency behavior
* Semiconductor devices
* Amplification and feedback
* Digital electronics

In the laboratory you will be able to:

* Use the oscilloscope, function generator, and various power supplies.
* Perform measurements and compare them with simulations.
* Follow the methodological cycle:

> Model → Simulation → Measurement → Interpretation

</div>

## 2. Important Clarifications

* **Attendance and active participation (10%)**
  Regular attendance, engagement in discussions, and contribution during lectures and seminars are evaluated.

* **Midterm exam (10%)**
  Written assessment covering the first part of the theoretical and conceptual material.

* **Projects (10%)**
  Individual or group projects focused on circuit analysis, simulation, design, or practical applications.

* **Laboratories (30%)**
  Evaluation is based on preparation, correct execution of experiments, data analysis, and final report.

* **Final exam (40%)**
  Written examination covering the entire course content (analog and digital).
