```python
import numpy as np
import matplotlib.pyplot as plt

# --- Component Values ---
R = 1000        # Ohms
C = 1e-6        # Farads

# --- Theoretical Calculations ---
fc_theory = 1/(2*np.pi*R*C)

print("Theoretical cutoff frequency (Hz):", fc_theory)
```

# 4. Procedure

1. Constructed circuit as shown.
2. Applied input signal of ____ Vpp.
3. Swept frequency from ____ to ____.

```python
# Example frequency sweep data
frequency = np.array([100, 300, 1000, 3000, 10000])
v_in = np.array([2, 2, 2, 2, 2])
v_out = np.array([1.98, 1.90, 1.40, 0.80, 0.20])

gain = v_out / v_in

df_data = pd.DataFrame({
    "Frequency (Hz)": frequency,
    "Vin (Vpp)": v_in,
    "Vout (Vpp)": v_out,
    "Gain": gain
})

df_data
```

```python
plt.semilogx(frequency, gain, 'o-', label="Experimental")
plt.axvline(fc_theory, linestyle='--', label="Theoretical fc")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Gain")
plt.title("Frequency Response")
plt.grid(True)
plt.legend()
plt.show()
```
