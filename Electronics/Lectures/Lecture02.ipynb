## 1. Why Study Sinusoidal Inputs?

Last week we studied the response to a step input.

Now we ask:

What happens if the input is a sinusoid?

Why is this important?

Because many real signals can be decomposed into sinusoids,
and circuits respond differently to different frequencies.

## 2. RC Circuit (Low-Pass Configuration)

$$ V_{in} \to R \to C \to Ground$$  
Output: $V_{out} = V_c$

Assume input:
$$ V_{in}(t) = V_0 \cos(\omega t)$$

## 3. Complex Impedance

For sinusoidal steady-state:

$$Z_R = R$$  
$$ Z_C = \frac{1}{(j\omega C)}$$

Voltage divider:

$$ \frac{V_{out}}{Vin} = \frac{Z_C}{(R + Z_C)} $$

$$ \boxed{H(\omega) = \frac{1}{(1 + j \omega RC)}} $$
This is the **transfer function**. Transfer function of a circuit = output/input voltage ratio in frequency domain.

## 4. Magnitude Response

$$ |H(\omega)| = \frac{1}{\sqrt{(1 + (\omega RC)^2)}} $$

* At low frequency $(\omega \to 0)$: $|H| \approx 1$, $ \phi=0$ → capacitor acts open → output ≈ input

* At high frequency $(\omega \to \infty)$: $|H| \to 0$, $\phi=-\arctan(\omega RC)$. → capacitor acts short → output small


## 5. Cutoff Frequency

Define cutoff $\omega_c$ such that: $|H(\omega_c)| = 1/\sqrt 2$. Solve: $\boxed{\omega_c = 1/RC}$. In frequency:

$$f_c = \frac{1}{(2\pi RC)}$$

```python
import numpy as np
import matplotlib.pyplot as plt

R = 1000
C = 1e-6
tau = R*C

f = np.logspace(1, 6, 1000)
omega = 2*np.pi*f

H = 1 / np.sqrt(1 + (omega*tau)**2)

plt.figure()
plt.semilogx(f, 20*np.log10(H))
plt.xlabel("Frequency [Hz]")
plt.ylabel("Magnitude [dB]")
plt.title("RC Low-Pass Bode Magnitude")
plt.grid(True, which='both')
plt.show()
```

## 6. Physical Interpretation

- The capacitor resists rapid voltage changes.
- High frequencies correspond to rapid changes.
- Therefore, high frequencies are attenuated.

This circuit acts as a low-pass filter.

## 8. Time vs Frequency Connection

Recall from Week 1:

τ = RC

Cutoff frequency:

fc = 1 / (2πRC)

Thus:

Slow time response ↔ Low cutoff frequency
Fast time response ↔ High cutoff frequency

Time and frequency are two views of the same system.
