# Instrumentet Laboratorike

## Objektivi

Përpara se të ndërtojmë ose të masim diçka, duhet të kuptojmë **çfarë janë realisht mjetet laboratorike**.

Një instrument laboratorik nuk është një kuti misterioze.  
Ai është thjesht një pajisje që:

- **Lidh** elementë elektrikë  
- **Mat** madhësi elektrike  
- **Gjeneron** sinjale elektrike  
- **Furnizon** energji elektrike të kontrolluar  

Nëse e kuptoni këtë klasifikim, laboratori bëhet i thjeshtë.


## Katër Kategoritë Themelore

| Kategoria   | Shembull mjeti        | Çfarë bën |
|------------|------------------------|-----------|
| Lidhje     | *Breadboard*           | Na lejon të ndërtojmë qarqe |
| Matje      | *Multimetër, Oshiloskop* | Vëzhgon madhësi elektrike |
| Gjenerim   | *Gjenerator Funksioni* | Prodhon sinjale prove të kontrolluara |
| Furnizim   | *Furnizim me Energji*  | Siguron energji elektrike të qëndrueshme |

Gjithçka që do të përdorim përshtatet brenda njërit prej këtyre katër roleve.


## Ideja Më e Rëndësishme

<div style="border:6px solid #007acc; background-color:#ddd; color:#000; border-radius: 5px; padding: 10px;">

Elektronika në laborator ndjek gjithmonë të njëjtën logjikë:

1. **Ndërto** një qark  
2. **Furnizoje** me energji  
3. **Stimuloje** me një sinjal  
4. **Mat** përgjigjen e tij  

**Ky është cikli i plotë eksperimental.**
</div>


## Çfarë Është Sinjali?

Sinjali është thjesht një **madhësi fizike që ndryshon me kohën**. Në elektronikë, sinjali zakonisht është:

- Tensioni
- Rryma

Në shumicën e rasteve ne vëzhgojmë: $V(t)$ tensionin si funksion i kohës.


## Mendësia Eksperimentale

Kur hyni në laborator, ju nuk jeni thjesht “duke përdorur pajisje”. Ju po i bëni pyetje sistemit (qarkut) duke pyetur:

- Çfarë ndodh nëse aplikoj këtë tension?
- Si reagon sistemi ndaj kësaj frekuence?
- A përputhet teoria me matjen?

**Instrumentet janë thjesht mjete që ju lejojnë të shihni përgjigjen.**


## Siguria dhe Disiplina

<div style="border:6px solid #007acc; background-color:#ddd; color:#000; border-radius: 5px; padding: 10px;">

Përpara se të prekni ndonjë instrument:

- **Mos supozoni kurrë lidhjet — verifikojini.**
- **Filloni gjithmonë me tension të ulët.**
- **Mendoni përpara se të kaloni në matjen e rrymës.**
- **Mos matni kurrë rezistencë në një qark të furnizuar.**

Praktika e mirë laboratorike është më e rëndësishme se shpejtësia.
</div>


Tani do ta shqyrtojmë secilin mjet veçmas:

1. Breadboard – ku ndërtohen qarqet  
2. Multimetër – si masim madhësi DC  
3. Furnizim me Energji – burim energjie i kontrolluar  
4. Gjenerator Funksioni – burim sinjali  
5. Oshiloskop – vizualizimi i tensionit kundrejt kohës  

Secili mjet do të bëhet i njohur dhe intuitiv.

**Ata nuk janë të komplikuar. Ata janë të saktë.**

***Dhe do të bëhen miqtë tuaj.***

# Pjesa 1 – Breadboard-i  
## Ku ndërtohen qarqet

<img src="../figs/Instruments/Wooden_Breadboard_Circuits.jpg" align="center" width=500 style="margin-left:auto; margin-right:auto"/>


## 1. Çfarë është breadboard-i?

Breadboard-i është një **platformë lidhjeje e përkohshme**. Ai na lejon të ndërtojmë qarqe **pa saldim**.  
Brenda breadboard-it ka shirita metalikë të fshehur që lidhin disa vrima mes tyre.  
Nëse i kuptoni këto lidhje të brendshme, e kuptoni breadboard-in.

<img src="../figs/Instruments/Ultrasound-PreAmp-Breadboard.jpg" align="center" width=500 style="margin-left:auto; margin-right:auto"/>


## 2. Struktura e brendshme

Një breadboard standard ka tri seksione kryesore:

1. **Rail-et e fuqisë (anësore)**
2. **Shiritat e terminaleve (zona qendrore)**
3. **Hapësira qendrore (gap)**


<img src="../figs/Instruments/breadboard.png" align="center" width=500 style="margin-left:auto; margin-right:auto"/>


### Rail-et e fuqisë

- Kolona të gjata vertikale në anë
- Zakonisht të shënuara me të kuqe (+) dhe blu (–)
- Përdoren për shpërndarjen e furnizimit (Vcc dhe GND)

⚠ E rëndësishme: Në shumë breadboard-e, rail-et e fuqisë janë të ndara në mes.  
Ato mund të mos jenë të lidhura nga sipër-poshtë.


### Shiritat e terminaleve (zona e punës)

- Rreshta të shkurtër horizontalë me 5 vrima
- Çdo rresht me 5 vrima është i lidhur brendshëm

Shembull: Nëse vendosni një tel në një vrimë të rreshtit, të gjitha 5 vrimat në atë rresht janë të lidhura elektrikisht.


### Hapësira qendrore (gap)

Hapësira në mes:

- Ndan anën e majtë nga ana e djathtë
- Lejon që qarqet e integruara (IC) të vendosen mbi hapësirë
- Dy anët e hapësirës NUK janë të lidhura


## 3. Modeli mendor

Mendoni breadboard-in si:

- Shumë “ura” të vogla metalike të fshehura nën plastikë
- Çdo rresht është një bus i vogël
- Rail-et anësore janë bus-e të gjata për furnizim

**Ai nuk krijon qark. Ai vetëm lidh komponentët.**


<div class="alert alert-info">
    
## 4. Shembulli i parë: Qark me LED

Do të ndërtojmë qarkun më të thjeshtë:

- Furnizim me energji
- Rezistencë
- LED

* Hapi 1 – Vendosni rezistencën:
  - Futni këmbët e rezistencës në dy rreshta të ndryshëm.
  - E rëndësishme: Të dyja këmbët NUK duhet të jenë në të njëjtin rresht. Përndryshe krijohet shkurtim.
* Hapi 2 – Vendosni LED-in
  - Këmba e gjatë → Ana pozitive
  - Këmba e shkurtër → Ana e tokës (GND) 
  Sërish: secila këmbë duhet të jetë në rresht të ndryshëm.
* Hapi 3 – Lidhni furnizimin
  - Lidhni + nga furnizimi te rezistenca
  - Lidhni LED-in me rail-in e tokës

Tani keni: Furnizim → Rezistencë → LED → Tokë

</div>


## 5. Gabime të zakonshme të fillestarëve

* **Gabimi 1: Lidhja në të njëjtin rresht:** Nëse të dyja këmbët e një komponenti janë në të njëjtin rresht, komponenti nuk bën asgjë.
* **Gabimi 2: Supozimi i rail-it:** Supozimi se rail-i anësor është i vazhdueshëm kur në fakt është i ndarë.
* **Gabimi 3: Lidhje të çrregullta:** Tela të gjatë krijojnë:
  - Konfuzion
  - Zhurmë
  - Probleme në diagnostikim
    Mbajini telat të shkurtër dhe të organizuar.

<img src="../figs/Instruments/real_circuit.png" align="center" width=500 style="margin-left:auto; margin-right:auto"/>


## 6. Praktikë e mirë në breadboard

- **Përdorni tela të kuq për pozitivin**
- **Përdorni të zi ose blu për tokën**
- **Mbani telat e sinjalit me ngjyra të ndryshme**
- **Ndërtoni nga e majta në të djathtë**
- **Kontrolloni lidhjet përpara furnizimit**

**Disiplina në lidhje ju kursen orë të tëra diagnostikimi.**


## 7. Çfarë NUK bën breadboard-i

Ai nuk:

- Kufizon rrymën
- Mbron komponentët
- Verifikon korrektësinë

Ai është vetëm një matricë lidhjesh. Ju jeni përgjegjës për projektimin.

## 8. Përmbledhje

Breadboard-i është:

- Një platformë lidhjeje e përkohshme
- Një matricë lidhjesh të fshehura
- Pika e nisjes për çdo qark laboratorik

Nëse e kuptoni strukturën e brendshme, rrallë do të bëni gabime në lidhje.

# Pjesa 2 – Multimetri  
## Mjeti juaj i parë i matjes


## 1. Çfarë është multimetri?

Multimetri është një pajisje që mat:

- Tensionin
- Rrymën
- Rezistencën

Ai kombinon disa instrumente në një mjet të vetëm.

Mendoni si:

- Voltmetër  
- Ampermetër  
- Ohmmetër  

Të gjitha në një pajisje të vetme.

## 2. Tre modalitetet më të rëndësishme të matjes

Për fillestarët, fokusohemi vetëm në tre modalitete:

1. **Tensioni (V)**
2. **Rryma (A ose mA)**
3. **Rezistenca (Ω)**

Nëse i kuptoni saktë këto tre, mund të kryeni me siguri shumicën e matjeve laboratorike.


### **A. Matja e tensionit** (modaliteti më i përdorur)

<div class="alert alert-success">

#### **Çfarë është tensioni?**

Tensioni është **diferenca e potencialit elektrik midis dy pikave**. <u>Tensionin e masim **midis dy nyjeve**.</u>

#### **Si lidhet multimetri**

- Vendosni rrotën te Tensioni DC (V⎓)
- Sonda e zezë → terminali COM
- Sonda e kuqe → terminali V
- Lidhni sondat **në paralel** me komponentin


Shumë e rëndësishme: Tensioni matet gjithmonë **në paralel**.  
Pse në paralel?  
Sepse multimetri në modalitetin e tensionit ka:

- Rezistencë të brendshme shumë të lartë
- Nuk e shqetëson në mënyrë domethënëse qarkun
</div>

<div class="alert alert-info">

#### Ushtrimet e para

Matni:

- Tensionin e një baterie
- Tensionin mbi një rezistencë
- Tensionin përpara (forward) të një LED-i

Pyetni veten:

- A është vlera e arsyeshme?
- A përputhet me pritjen teorike?

</div>


### B. Matja e rrymës (zona e rrezikut)

Këtu bëhen shumica e gabimeve të fillestarëve.


<div class="alert alert-success">

#### Çfarë është rryma?

**Rryma është rrjedhja e ngarkesës nëpër një komponent.** Ndryshe nga tensioni, rryma rrjedh *nëpër* një degë.

#### Si lidhet multimetri

- Zhvendosni sondën e kuqe te terminali A ose mA
- Vendosni rrotën te rryma DC
- Ndërpritni qarkun
- <u>Futeni multimetri **në seri**</u>

<u>Shumë e rëndësishme: Rryma matet gjithmonë **në seri**.</u>


#### Pse është e rrezikshme?

Në modalitetin e rrymës, multimetri ka: rezistencë të brendshme shumë të ulët.  
Nëse e lidhni në paralel si voltmetër: krijoni një shkurtim.

Pasojat e mundshme:

- Digjet siguresa e brendshme e multimetrit
- Dëmtohet furnizimi me energji
- Dëmtohen komponentët


#### Procedurë e sigurt

Përpara se të matni rrymën:

1. Fikni furnizimin
2. Rikonfiguroni qarkun
3. Kontrolloni përsëri pozicionin e sondave
4. Pastaj ndizni furnizimin

Mos nxitoni kurrë me matjen e rrymës.

</div>



### C. Matja e rezistencës (modaliteti Ω)

<div class="alert alert-success">

#### Çfarë është rezistenca?

Rezistenca përshkruan sa e pengon një komponent rrjedhjen e rrymës.


#### Rregull shumë i rëndësishëm

Mos matni kurrë rezistencë në një qark të furnizuar.

Pse?

Sepse multimetri injekton një rrymë prove të vogël të brendshme.

Nëse ka tension të jashtëm:

- Matja del e pasaktë
- Multimetri mund të dëmtohet


#### Si matet

- Fikni furnizimin e qarkut (Power OFF)
- Vendosni rrotën te Ω
- Matni direkt në paralel mbi komponentin

</div>

<div class="alert alert-info">

#### Ushtrimet e para

Matni:

- Një rezistencë të lirë (jo të lidhur)
- Krahasoni vlerën e matur me kodin e ngjyrave
- Matni rezistencën midis dy nyjeve në breadboard për të verifikuar lidhjet

Kjo është gjithashtu e dobishme për diagnostikimin e lidhjeve.

</div>



## 3. Rikujtesë

| Modaliteti      | Rezistenca e brendshme | Lloji i lidhjes |
|-----------|-------------------------|----------------|
| Tensioni   | Shumë e lartë          | **Paralel**       |
| Rryma     | Shumë e ulët           | **Seri**         |
| Rezistenca| Burim i brendshëm      | **Vetëm me qark OFF** |

Nëse e mbani mend këtë tabelë, shmangni 90% të gabimeve.


## 4. Gabime të zakonshme të fillestarëve

- **Harrimi i kthimit të sondës nga A në V pas matjes së rrymës**
- **Matja e rrymës në paralel**
- **Matja e rezistencës në qark të furnizuar**
- Mos kontrollimi i intervalit (range) të matjes


## 5. Zakon i mirë laboratorik

Përpara matjes, pyesni veten:

- Cilën madhësi po mas?
- Ku “jeton” ajo në qark?
- Paralel apo seri?

Matja nuk është rastësi. Ajo është e qëllimshme.


## 6. Përmbledhje

Multimetri është:

- Instrumenti më i përdorur në laborator
- I thjeshtë por i fuqishëm
- I sigurt nëse përdoret saktë

Rregullat e arta:

* Tensioni → Paralel  
* Rryma → Seri  
* Rezistenca → Power OFF  

Zotërojeni këtë mjet dhe laboratori bëhet i parashikueshëm.

# Pjesa 3 – Furnizimi Laboratorik me Energji  
## Energji e kontrolluar për qarkun tuaj


<img src="../figs/Instruments/PSU.jpg" align="center" width=500 style="margin-left:auto; margin-right:auto"/>


## 1. Çfarë është furnizimi me energji?

Një furnizim laboratorik me energji siguron:
- Tension të qëndrueshëm
- Rrymë të kontrolluar
- Dalje të rregullueshme

Mendoni si: një **bateri inteligjente** me mbrojtje.

Ndryshe nga një bateri, ju mund:
- Të rregulloni tensionin
- Të kufizoni rrymën
- Të monitoroni konsumin


## 2. Tre kontrollet kryesore

Shumica e furnizimeve laboratorike me energji ju lejojnë të vendosni:
1. Tensionin (V)
2. Kufirin e rrymës (A)
3. Daljen ON/OFF

Këto janë të vetmet kontrolle që fillestarët duhet t’i kuptojnë në fillim.


## 3. Vendosja e tensionit

Butoni i tensionit vendos tensionin e dëshiruar në dalje.

Shembull:
- 5 V për qarqe digjitale
- 9–12 V për eksperimente të vogla analoge

Furnizimi do të përpiqet ta mbajë këtë tension.


## 4. Kufiri i rrymës (shumë i rëndësishëm)

Kufiri i rrymës përcakton: rrymën maksimale që furnizimi do të lejojë të rrjedhë. Kjo është mbrojtje.

Nëse qarku juaj përpiqet të tërheqë shumë rrymë:
- Furnizimi kalon në modalitetin e kufizimit të rrymës
- Tensioni bie automatikisht
- Komponentët mbrohen

Për këtë arsye furnizimi është më i sigurt se bateria.


## 5. Dy modalitete pune

Shumica e furnizimeve punojnë në dy modalitete:

### CV – Tension konstant

- Tensioni është i qëndrueshëm
- Rryma ndryshon në varësi të ngarkesës

Ky është funksionimi normal.

### CC – Rrymë konstante

- Kufiri i rrymës është arritur
- Tensioni reduktohet automatikisht

Kjo ndodh nëse:
- Ka një shkurtim
- Qarku tërheq shumë rrymë


## 6. Procedura e sigurt e nisjes

Përpara se të lidhni qarkun:

1. Vendosni daljen OFF
2. Vendosni tensionin në vlerën e dëshiruar
3. Vendosni kufirin e rrymës në një vlerë të ulët dhe të sigurt
4. Lidhni qarkun
5. Vendosni daljen ON

Mos e lidhni kurrë një qark me një dalje aktive me tension të panjohur.


## 7. Zgjedhja e një kufiri të sigurt të rrymës

Për qarqe të vogla fillestare:

- LED → maksimumi 20 mA
- Qarqe të vogla në breadboard → 100–200 mA si kufi i sigurt

Nëse nuk jeni të sigurt:
Filloni me vlera të ulëta. Rriteni gradualisht. Mbrojtja është më e mirë se zëvendësimi.


## 8. Eksperiment demonstrues

Shkurtëzoni për pak çaste terminalet e daljes.  
Vëzhgoni:
- Ekrani kalon në CC
- Tensioni bie
- Rryma mbetet e kufizuar

Kjo tregon si funksionon mbrojtja. Mos e mbani të shkurtruar për shumë kohë.


## 9. Gabime të zakonshme të fillestarëve

- **Harrimi i vendosjes së kufirit të rrymës**
- **Lënia e tensionit të lartë nga përdoruesi i mëparshëm**
- **Lidhja e qarkut pa kontrolluar parametrat**
- **Supozimi se furnizimi është i fikur**

Kontrolloni gjithmonë ekranin.


## 10. Çfarë NUK bën furnizimi me energji

Ai nuk:
- **Parandalon gabimet e polaritetit të kundërt**
- **Korrigjon automatikisht lidhjet e gabuara**
- **Zbulon të gjitha defektet**

**Ju jeni përgjegjës për lidhjet e sakta.**


## 11. Përmbledhje

Furnizimi me energji është:
- Burim energjie i kontrollueshëm
- Pajisje mbrojtëse
- Mjet diagnostikimi

Zakonet kryesore:
- **Vendosni tensionin**
- **Vendosni kufirin e rrymës**
- **Verifikoni përpara furnizimit**
- **Mendoni përpara se të ndizni**

**Furnizimi me energji është hapi i fundit, jo i pari.**

# Pjesa 4 – Gjeneratori i Funksionit  
## Krijimi i sinjaleve për të testuar qarkun tuaj

<img src="../figs/Instruments/Siglent_SDG1025.jpg" align="center" width=500 style="margin-left:auto; margin-right:auto"/>


## 1. Çfarë është gjeneratori i funksionit?

Gjeneratori i funksionit është një pajisje që prodhon **sinjale elektrike**.  
Në vend të një tensioni konstant (si furnizimi me energji), ai gjeneron tensione që **ndryshojnë me kohën**.

Mendoni për të si një makinë që i bën qarkut tuaj një pyetje.  
<u> **Nëse furnizimi me energji jep energji, gjeneratori i funksionit jep stimulim.** </u>


<div style="border:6px solid #007acc; background-color:#ddd; color:#000; border-radius: 5px; padding: 10px;">

## 2. Çfarë është sinjali?

Sinjali është një tension që ndryshon me kohën: $V(t)$  
Format më të zakonshme të sinjalit:
- Valë sinusoidale
- Valë katrore
- Valë trekëndore

</div>


## 3. Tre kontrollet më të rëndësishme

Fillestarët duhet të kuptojnë vetëm tre parametra:

1. Frekuenca
2. Amplituda
3. Offset

Gjithçka tjetër është përpunim i mëtejshëm.


### A. Frekuenca

Frekuenca na tregon sa shpejt oscilon sinjali.  
Matet në:
- Hertz (Hz)
- kHz (10³ Hz)
- MHz (10⁶ Hz)

Frekuencë e ulët → sinjal i ngadaltë  
Frekuencë e lartë → sinjal i shpejtë


### B. Amplituda

Amplituda na tregon sa e madhe është luhatja e tensionit.  
Ajo mund të shprehet si:
- Tension kulmor (peak)
- Tension kulm-më-kulm (Vpp)


### C. Offset

Offset e zhvendos sinjalin lart ose poshtë.

Shembull:
- 0 V offset → sinjali i qendruar rreth zeros
- +2 V offset → sinjali i zhvendosur lart

Kjo është e rëndësishme kur qarqet nuk mund të pranojnë tensione negative.


## 4. Lidhja e daljes

Dalja zakonisht është:
- Konektor BNC
- E referuar ndaj tokës

Zakonisht lidhet:
- Pini qendror → sinjali
- Mburoja e jashtme → toka


## 5. Koncept i rëndësishëm: Dalja 50 Ω

Shumica e gjeneratorëve të funksionit kanë: impedancë të brendshme dalëse 50 Ω.  
**Kjo do të thotë: amplituda e shfaqur është e saktë kur lidhet me një ngarkesë 50 Ω.**  
**Nëse lidhet me impedancë të lartë (si një oshiloskop): tensioni i matur mund të duket i dyfishuar.**

Mbani mend: verifikoni gjithmonë amplitudën në oshiloskop.

## 6. Eksperimenti i parë

## 2. Çfarë shfaq?

Ekrani shfaq:
* Boshti vertikal → Tension  
* Boshti horizontal → Kohë  

Pra, ajo që shihni është: $V(t)$. Kjo është ideja më e rëndësishme.


<div style="border:6px solid #007acc; background-color:#ddd; color:#000; border-radius: 5px; padding: 10px;">

## 3. Katër kontrollet thelbësore

Fillestarët duhet të zotërojnë vetëm katër kontrolle:

1. Shkalla vertikale (Volts/div)
2. Baza kohore (Time/div)
3. Trigger
4. Zbutja e sondës (1x / 10x)

Në fillim, injoroni gjithçka tjetër.


### A. Shkalla vertikale (Volt për ndarje)

**Kjo përcakton sa tension përfaqëson çdo katror vertikal.**  
Nëse vendoset 1 V/div, çdo katror vertikal është 1 volt.

Nëse sinjali zë 4 katrorë ⇨ Tensioni ≈ 4 V.

**Rregulloni derisa vala të përshtatet mirë në ekran.**


### B. Baza kohore (Kohë për ndarje)

**Kjo përcakton sa kohë përfaqëson çdo katror horizontal.**  
Nëse vendoset 1 ms/div, çdo katror horizontal është 1 milisekond.

Nëse një cikël i plotë zë 5 ndarje ⇨ Perioda = 5 ms & Frekuenca = 1/Perioda


### C. Trigger (kontrolli i stabilitetit)

**Pa trigger të rregullt, vala lëviz në mënyrë të çrregullt.**  
Trigger-i:
- **Stabilizon valën**
- **E lidh shfaqjen me një pikë reference**

Vendosje bazë:
- **Burimi i trigger-it → Kanali 1**
- **Modaliteti i trigger-it → Auto ose Normal**
- **Niveli i trigger-it → Rregulloni derisa vala të jetë stabile**

Nëse sinjali lëviz, rregulloni trigger level.


### D. Zbutja e sondës (1x kundrejt 10x)

Shumica e sondave kanë një çelës të vogël 1x ose 10x.

Modaliteti 10x:
- **E zvogëlon sinjalin me faktor 10**
- **Rrit impedancën e hyrjes**
- **Zvogëlon efektin e ngarkimit (loading)**

**Kontrolloni gjithmonë që vendosja e sondës përputhet me vendosjen e kanalit në oshiloskop. Mospërputhje → lexim i gabuar i amplitudës.**

## 7. AC vs DC Coupling

Shumica e oshiloskopëve lejojnë:
* DC Coupling → Shfaq sinjalin e plotë (përfshirë offset-in)  
* AC Coupling → Heq offset-in DC

**Përdorni AC coupling kur dëshironi të vëzhgoni sinjale të vogla AC mbi vlera të mëdha DC.**

Kjo ndryshon mënyrën si mendoni për elektronikën.
