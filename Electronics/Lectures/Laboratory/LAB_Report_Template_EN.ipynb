# 4. Measurement Data

```python
import pandas as pd

data = pd.DataFrame({
    "Frequency (Hz)": [],
    "Vin (Vpp)": [],
    "Vout (Vpp)": []
})

data
```

```python
data["Gain"] = data["Vout (Vpp)"] / data["Vin (Vpp)"]
data
```

# 5. Data Analysis

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(6,4))
plt.semilogx(data["Frequency (Hz)"], 20*np.log10(data["Gain"]))
plt.xlabel("Frequency (Hz)")
plt.ylabel("Gain (dB)")
plt.title("Bode Magnitude Plot")
plt.grid(True, which="both")
plt.show()
```

From the plot determine:

- Experimental cutoff frequency  
- Slope after cutoff  
- Agreement with theoretical prediction
