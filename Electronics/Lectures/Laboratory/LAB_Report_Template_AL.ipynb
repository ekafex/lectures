# 4. Të Dhënat e Matjes

```python
import pandas as pd

data = pd.DataFrame({
    "Frekuenca (Hz)": [],
    "Vin (Vpp)": [],
    "Vout (Vpp)": []
})

data
```

```python
data["Fitimi"] = data["Vout (Vpp)"] / data["Vin (Vpp)"]
data
```

# 5. Analiza e të Dhënave

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(6,4))
plt.semilogx(data["Frekuenca (Hz)"], 20*np.log10(data["Fitimi"]))
plt.xlabel("Frekuenca (Hz)")
plt.ylabel("Fitimi (dB)")
plt.title("Grafiku Bode (Amplituda)")
plt.grid(True, which="both")
plt.show()
```

Nga grafiku përcaktoni:

- Frekuencën e prerjes (eksperimentale)  
- Pjerrësinë pas frekuencës së prerjes  
- Krahasimin me vlerën teorike
