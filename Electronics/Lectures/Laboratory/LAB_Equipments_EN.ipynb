# Laboratory Instruments

## Objective

Before building or measuring anything, we must understand **what laboratory tools really are**.

A laboratory instrument is not a mysterious box.  
It is simply a device that either:

- **Connects** electrical elements  
- **Measures** electrical quantities  
- **Generates** electrical signals  
- **Supplies** controlled electrical energy  

If you understand this classification, the laboratory becomes simple.


## The Four Fundamental Categories

| Category   | Tool Example        | What It Does |
|------------|--------------------|--------------|
| Connect    | *Breadboard*          | Allows us to build circuits |
| Measure    | *Multimeter, Oscilloscope* | Observes electrical quantities |
| Generate   | *Function Generator*  | Produces controlled test signals |
| Supply     | *Power Supply*        | Provides stable electrical energy |

Everything we will use fits into one of these four roles.


## The Most Important Idea

<div style="border:6px solid #007acc; background-color:#ddd; color:#000; border-radius: 5px; padding: 10px;">

Electronics in the laboratory always follows the same logic:

1. **Build** a circuit  
2. **Supply** it with energy  
3. **Stimulate** it with a signal  
4. **Measure** its response  

**That is the complete experimental cycle.**
</div>


## What Is a Signal?

A signal is simply a **physical quantity that changes with time**. In electronics, the signal is usually:

- Voltage
- Current

Most of the time we observe: $V(t)$ Voltage as a function of time.


## The Experimental Mindset

When you enter the laboratory, you are not “using devices”. You are quering the system (circuit) by asking:

- What happens if I apply this voltage?
- How does the system respond to this frequency?
- Does theory match measurement?

**The instruments are just tools that allow you to see the answer.**


## Safety and Discipline

<div style="border:6px solid #007acc; background-color:#ddd; color:#000; border-radius: 5px; padding: 10px;">

Before touching any instrument:

- **Never assume connections — verify them**.
- **Always start with low voltage.**
- **Think before switching to current measurement.**
- **Never measure resistance in a powered circuit.**

Good laboratory practice is more important than speed.
</div>


We will now examine each tool individually:

1. Breadboard – where circuits are built  
2. Multimeter – how we measure DC quantities  
3. Power Supply – controlled energy source  
4. Function Generator – signal source  
5. Oscilloscope – visualizing voltage vs time  

Each tool will become familiar and intuitive.

**They are not complicated. They are precise.**

***And they will become your friends.***

# Part 1 – The Breadboard  
## Where Circuits Are Built

<img src="../figs/Instruments/Wooden_Breadboard_Circuits.jpg" align="center" width=500 style="margin-left:auto; margin-right:auto"/>


## 1. What Is a Breadboard?

A breadboard is a **temporary wiring platform**. It allows us to build circuits **without soldering**.
Inside the breadboard there are hidden metal strips that connect certain holes together.  
If you understand these internal connections, you understand the breadboard.

<img src="../figs/Instruments/Ultrasound-PreAmp-Breadboard.jpg" align="center" width=500 style="margin-left:auto; margin-right:auto"/>


## 2. The Internal Structure

A standard breadboard has three main sections:

1. **Power Rails (Sides)**
2. **Terminal Strips (Middle Area)**
3. **Central Gap**


<img src="../figs/Instruments/breadboard.png" align="center" width=500 style="margin-left:auto; margin-right:auto"/>


### Power Rails

- Long vertical columns on the sides
- Usually marked with red (+) and blue (–)
- Used to distribute power (Vcc and GND)

⚠ Important: On many breadboards, the power rails are split in the middle.  
They may not be connected from top to bottom.


### Terminal Strips (Working Area)

- Short horizontal rows of 5 holes
- Each row of 5 holes is internally connected

Example: If you place a wire in one hole of a row, all 5 holes in that row are electrically connected.


### The Central Gap

The gap in the middle:

- Separates the left and right sides
- Allows integrated circuits (ICs) to sit across the gap
- The two sides of the gap are NOT connected


## 3. Mental Model

Think of the breadboard as:

- Many small metal bridges hidden under plastic
- Each row is a small bus
- The side rails are long buses for power

**It does not create a circuit. It only connects components.**


<div class="alert alert-info">
    
## 4. First Example: LED Circuit

We will build the simplest circuit:

- Power supply
- Resistor
- LED

* Step 1 – Place the Resistor:
  - Insert resistor legs into two different rows.
  - Important: Both legs must NOT be in the same row. Otherwise it is shorted.
* Step 2 – Place the LED
  - Long leg → Positive side
  - Short leg → Ground side 
  Again: each leg must be in a different row.
* Step 3 – Connect Power
  - Connect + from power supply to resistor
  - Connect LED to ground rail

Now you have: Power → Resistor → LED → Ground

</div>


## 5. Common Beginner Mistakes

* **Mistake 1: Same Row Connection:** If both legs of a component are in the same row,the component does nothing.
* **Mistake 2: Rail Assumption:** Assuming the side rail is continuous when it is split.
* **Mistake 3: Messy Wiring:** Long wires create:
  - Confusion
  - Noise
  - Debugging problems
    Keep wires short and organized.

<img src="../figs/Instruments/real_circuit.png" align="center" width=500 style="margin-left:auto; margin-right:auto"/>


## 6. Good Breadboard Practice

- **Use red wires for positive**
- **Use black or blue wires for ground**
- **Keep signal wires different colors**
- **Build from left to right**
- **Check connections before powering**

**Discipline in wiring saves hours of debugging.**


## 7. What the Breadboard Does NOT Do

It does not:

- Limit current
- Protect components
- Check correctness

It is only a connection matrix. You are responsible for the design.

## 8. Summary

The breadboard is:

- A temporary wiring platform
- A matrix of hidden connections
- The starting point of every laboratory circuit

If you understand its internal structure, you will rarely make wiring mistakes.

## 2. The Three Most Important Measurement Modes

For beginners, we focus on three modes only:

1. **Voltage (V)**
2. **Current (A or mA)**
3. **Resistance (Ω)**

If you understand these three properly, you can safely perform most laboratory measurements.


### **A. Measuring Voltage** (Most Common Mode)

<div class="alert alert-success">

#### **What Is Voltage?**

Voltage is a **difference in electric potential between two points**. <u>We measure voltage **between two nodes**.</u>

#### **How to Connect the Multimeter**

- Set the dial to DC Voltage (V⎓)
- Black probe → COM terminal
- Red probe → V terminal
- Connect probes **in parallel** with the component


Very important: Voltage is always measured **in parallel**.
Why Parallel?
Because the multimeter in voltage mode has:

- Very high internal resistance
- It does not significantly disturb the circuit
</div>

<div class="alert alert-info">

#### First Exercises

Measure:

- A battery voltage
- The voltage across a resistor
- The forward voltage of an LED

Ask yourself:

- Is the value reasonable?
- Does it match theoretical expectation?

</div>


### B. Measuring Current (Danger Zone)

This is where beginners make mistakes.


<div class="alert alert-success">

#### What Is Current?

**Current is the flow of charge through a component.** Unlike voltage, current flows *through* a branch.

#### How to Connect the Multimeter

- Move red probe to the A or mA terminal
- Set dial to DC current
- Break the circuit
- <u>Insert the multimeter **in series**</u>

<u>Very important: Current is always measured **in series**.</u>


#### Why Is This Dangerous?

In current mode, the multimeter has: Very low internal resistance
If you connect it in parallel like a voltmeter: You create a short circuit.

Possible consequences:

- Blown fuse inside multimeter
- Damaged power supply
- Burned components


#### Safe Procedure

Before measuring current:

1. Turn off power
2. Reconfigure the circuit
3. Double-check probe position
4. Then power on

Never rush current measurements.

</div>



### C. Measuring Resistance (Ω Mode)

<div class="alert alert-success">

#### What Is Resistance?

Resistance describes how much a component opposes current.


#### Very Important Rule

Never measure resistance in a powered circuit.

Why?

Because the multimeter injects a small internal test current.

If external voltage is present:

- Measurement becomes incorrect
- The meter may be damaged


#### How to Measure

- Power OFF the circuit
- Set dial to Ω
- Measure directly across the component

</div>

<div class="alert alert-info">

#### First Exercises

Measure:

- A loose resistor (not connected)
- Compare measured value to color code
- Measure resistance between two breadboard nodes to verify connections

This is also useful to debug wiring.

</div>



## 3. Reminder

| Mode      | Internal Resistance | Connection Type |
|-----------|--------------------|----------------|
| Voltage   | Very High          | **Parallel**       |
| Current   | Very Low           | **Series**         |
| Resistance| Internal source    | **Power OFF only** |

If you remember this table, you avoid 90% of mistakes.


## 4. Common Beginner Errors

- **Forgetting to move probe back from A to V after current measurement**
- **Measuring current in parallel**
- **Measuring resistance on powered circuit**
- Not checking measurement range


## 5. Good Laboratory Habit

Before measuring, ask yourself:

- What quantity am I measuring?
- Where does it live in the circuit?
- Parallel or series?

Measurement is not random. It is deliberate.


## 6. Summary

The multimeter is:

- The most used instrument in the lab
- Simple but powerful
- Safe if used correctly

Golden rules:

* Voltage → Parallel  
* Current → Series  
* Resistance → Power OFF  

Master this tool, and the laboratory becomes predictable.

# Part 3 – The Laboratory Power Supply  
## Controlled Energy for Your Circuit


<img src="../figs/Instruments/PSU.jpg" align="center" width=500 style="margin-left:auto; margin-right:auto"/>


## 1. What Is a Power Supply?

A laboratory power supply provides:
- A stable voltage
- A controlled current
- Adjustable output

Think of it as: A **smart battery** with protection.

Unlike a battery, you can:
- Adjust voltage
- Limit current
- Monitor consumption


## 2. The Three Main Controls

Most laboratory power supplies allow you to set:
1. Voltage (V)
2. Current Limit (A)
3. Output ON/OFF

These are the only controls beginners must understand at first.


## 3. Voltage Setting

The voltage knob sets the desired output voltage.

Example:
- 5 V for digital circuits
- 9–12 V for small analog experiments

The power supply will try to maintain this voltage.


## 4. Current Limit (Very Important)

The current limit defines: The maximum current the supply will allow to flow. This is protection.

If your circuit tries to draw too much current:
- The supply switches to current-limited mode
- Voltage drops automatically
- Components are protected

This is why the power supply is safer than a battery.


## 5. Two Operating Modes

Most supplies operate in two modes:

### CV – Constant Voltage

- Voltage is stable
- Current varies depending on load

This is normal operation.

### CC – Constant Current

- Current limit has been reached
- Voltage is reduced automatically

This happens if:
- There is a short circuit
- The circuit draws too much current


## 6. Safe Startup Procedure

Before connecting your circuit:

1. Turn output OFF
2. Set voltage to desired value
3. Set current limit to a safe low value
4. Connect the circuit
5. Turn output ON

Never connect a circuit to an active unknown voltage.


## 7. Choosing a Safe Current Limit

For small beginner circuits:

- LEDs → 20 mA max
- Small breadboard circuits → 100–200 mA safe limit

If unsure:
Start low. Increase slowly. Protection is better than replacement.


## 8. Demonstration Experiment

Short the output terminals briefly.
Observe:
- The display switches to CC
- Voltage drops
- Current stays limited

This shows how protection works. Do not keep it shorted for long.


## 9. Common Beginner Mistakes

- **Forgetting to set current limit**
- **Leaving voltage high from previous user**
- **Connecting circuit before checking settings**
- **Assuming the supply is off**

Always check the display.


## 10. What the Power Supply Does NOT Do

It does not:
- **Prevent reversed polarity mistakes**
- **Automatically correct wrong wiring**
- **Detect all faults**

**You are responsible for correct connections.**


## 11. Summary

The power supply is:
- A controllable energy source
- A protective device
- A diagnostic tool

Golden habits:
- **Set voltage**
- **Set current limit**
- **Verify before powering**
- **Think before switching ON**

**Power is the last step, not the first.**

# Part 4 – The Function Generator  
## Creating Signals to Test Your Circuit

<img src="../figs/Instruments/Siglent_SDG1025.jpg" align="center" width=500 style="margin-left:auto; margin-right:auto"/>


## 1. What Is a Function Generator?

A function generator is a device that produces **electrical signals**.
Instead of a constant voltage (like a power supply), it generates voltages that **change with time**.

Think of it as a machine that asks your circuit a question.
<u> **If the power supply gives energy, the function generator gives stimulation.** </u>


<div style="border:6px solid #007acc; background-color:#ddd; color:#000; border-radius: 5px; padding: 10px;">

## 2. What Is a Signal?

A signal is a voltage that varies with time: $V(t)$
Most common signal shapes:
- Sine wave
- Square wave
- Triangle wave

</div>


## 3. The Three Most Important Controls

Beginners must understand only three parameters:

1. Frequency
2. Amplitude
3. Offset

Everything else is refinement.


### A. Frequency

Frequency tells us how fast the signal oscillates.
Measured in:
- Hertz (Hz)
- kHz (10³ Hz)
- MHz (10⁶ Hz)

Low frequency → slow signal  
High frequency → fast signal


### B. Amplitude

Amplitude tells us how large the voltage swing is.
It can be expressed as:
- Peak voltage
- Peak-to-peak voltage (Vpp)


### C. Offset

Offset shifts the signal up or down.

Example:
- 0 V offset → signal centered around 0
- +2 V offset → signal shifted upward

This is important when circuits cannot accept negative voltages.


## 4. Output Connection

The output is usually:
- A BNC connector
- Ground referenced

You typically connect:
- Center pin → signal
- Outer shield → ground


## 5. Important Concept: The 50 Ω Output

Most function generators have: Internal output impedance of 50 Ω.
**This means: The displayed amplitude is accurate when connected to a 50 Ω load.**
**If connected to high impedance (like an oscilloscope): The measured voltage may appear doubled.**

Just remember Always verify the amplitude on the oscilloscope.


<div class="alert alert-success">

## 6. First Experiment
1. Set sine wave
2. Frequency = 1 kHz
3. Amplitude = 2 Vpp
4. Offset = 0 V

Connect to oscilloscope.
Observe:
- Shape
- Period
- Peak-to-peak value

Then change frequency and observe effect.

</div>


## 7. What Is the Purpose of the Function Generator?

We use it to:
- Test circuits
- Study frequency response
- Observe time response
- Simulate real-world signals

Examples:
- Apply sine wave to RC circuit → observe filtering
- Apply square wave → observe rise time



## 8. Common Beginner Mistakes

- **Confusing amplitude with offset**
- **Forgetting ground connection**
- **Ignoring output impedance**
- **Setting too large amplitude**

**Always start with small signals.**


## 9. Summary

The function generator:

- Produces time-varying voltage
- Allows control of frequency and amplitude
- Is used to stimulate circuits

Golden parameters:
* Frequency  
* Amplitude  
* Offset  

**Control the input → observe the output.**

The oscilloscope will allow us to see what happens next.

# Part 5 – The Oscilloscope  
## Seeing Voltage as a Function of Time

<img src="../figs/Instruments/Oscilloscope_Clean.svg" align="center" width=900 style="margin-left:auto; margin-right:auto"/>

## 1. What Is an Oscilloscope?

**An oscilloscope displays Voltage versus Time. That is all!**
It allows you to *see* electrical signals.
**If the multimeter gives a number, the oscilloscope shows behavior.**

<img src="../figs/Instruments/analog.jpg" align="center" width=900 style="margin-left:auto; margin-right:auto"/> 

<img src="../figs/Instruments/siglent_sds1072cml.jpg" align="center" width=900 style="margin-left:auto; margin-right:auto"/>


## 2. What Does It Display?

The screen shows:
* Vertical axis → Voltage  
* Horizontal axis → Time  

So what you see is: $V(t)$. This is the most important idea.


<div style="border:6px solid #007acc; background-color:#ddd; color:#000; border-radius: 5px; padding: 10px;">

## 3. The Four Essential Controls

Beginners must master only four controls:

1. Vertical Scale (Volts/div)
2. Time Base (Time/div)
3. Trigger
4. Probe Attenuation (1x / 10x)

Ignore everything else at first.


### A. Vertical Scale (Volts per Division)

**This sets how much voltage each vertical square represents.**
If set to 1 V/div, Each vertical square equals 1 volt.

If the signal occupies 4 squares ⇨ Voltage ≈ 4 V.

**Adjust until the waveform fits nicely on screen.**


### B. Time Base (Time per Division)

**This sets how much time each horizontal square represents.**
If set to 1 ms/div, each horizontal square equals 1 millisecond.

If one full cycle spans 5 divisions ⇨ Period = 5 ms & Frequency = 1/Period


### C. Trigger (Stability Control)

**Without proper triggering, the waveform moves randomly.**
The trigger:
- **Stabilizes the waveform**
- **Locks the display to a reference point**

Basic setup:
- **Trigger source → Channel 1**
- **Trigger mode → Auto or Normal**
- **Trigger level → Adjust until stable**

If the signal moves, adjust trigger level.


### D. Probe Attenuation (1x vs 10x)

Most probes have a small switch 1x or 10x.

10x mode:
- **Reduces signal by factor of 10**
- **Increases input impedance**
- **Reduces loading effect**

**Always check that probe setting matches oscilloscope channel setting. Mismatch → incorrect amplitude reading.**

</div>

<div style="border:6px solid #007acc; background-color:#ddd; color:#000; border-radius: 5px; padding: 10px;">

## 4. Connecting the Oscilloscope

<img src="../figs/Instruments/probes.jpg" align="center" width=500 style="margin-left:auto; margin-right:auto"/>

Important. The probe has two parts:
- Tip → Signal
- Ground clip → Ground

**Always connect ground clip to circuit ground. Never connect the ground clip to a high-voltage node.**

**The oscilloscope ground is usually earth-referenced. Wrong grounding can cause short circuits.**

</div>

<div class="alert alert-success">

## 5. First Experiment

Connect function generator:
- Sine wave
- 1 kHz
- 2 Vpp
- 0 V offset

Steps:
1. Adjust Volts/div
2. Adjust Time/div
3. Adjust trigger
4. Measure peak-to-peak voltage
5. Measure period

Then: 
* Change frequency and observe.
* Change amplitude and observe.

**Learning happens by adjusting.**

</div>


## 6. What You Can Measure

With an oscilloscope you can determine:

- **Amplitude**
- **Peak-to-peak voltage**
- **Period**
- **Frequency**
- **Rise time**
- **DC offset**

**It transforms abstract signals into visible reality.**


## 7. AC vs DC Coupling

Most oscilloscopes allow:
* DC Coupling → Shows full signal (including offset)  
* AC Coupling → Removes DC offset

**Use AC coupling when you want to observe small AC signals on top of large DC values.**


## 8. Common Beginner Mistakes

- **Forgetting to connect ground**
- **Using wrong probe attenuation**
- **Ignoring trigger settings**
- **Setting time scale too fast or too slow**
- **Measuring without checking units**

**Always look at, *Volts/div & Time/div*, before interpreting the waveform.**


## 9. Why the Oscilloscope Is Important

**The oscilloscope teaches signals are dynamic.**
Circuits are not static numbers. They evolve in time.
This changes how you think about electronics.


## 10. Summary

The oscilloscope is:
- A voltage vs time viewer
- A diagnostic tool
- The most powerful instrument in the lab

Master:
* Volts/div  
* Time/div  
* Trigger  
* Probe mode  

**Once you understand these, you can analyze almost any basic circuit.**
