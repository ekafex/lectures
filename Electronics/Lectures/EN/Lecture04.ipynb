## 1. Why Study Electronics?

<u>Electronics is the science and technology that allows us to **measure, process, transmit, and control signals**.</u>

**Almost every modern scientific instrument relies on electronics.**

Examples include:
* scientific detectors
* communication systems
* computers
* sensors
* control systems
* measurement equipment

In physics laboratories, electronics is essential because most physical phenomena must first be **converted into electrical signals** before they can be measured or analyzed.

Examples include:
* sound pressure in air
* light intensity from a star
* temperature of a material
* magnetic field near a wire
* motion of a mechanical system

To analyze these quantities we represent them as **signals**.

<u>**A signal is simply a physical quantity expressed as a function of time.**</u>

Examples:
* Voltage signal $V(t)$
* Current signal $I(t)$
* Light intensity signal $I_{\rm light}(t)$
* Sound pressure signal $p(t)$
* etc ...

Electronics allows us to **convert many physical quantities into electrical signals**, because electrical signals are easy to manipulate, amplify, and measure.


### Why Electrical Signals?

**Electrical signals are extremely convenient because they can be easily manipulated.**
Using electronic circuits we can:
* **amplify weak signals**
* **remove noise**
* **select specific frequencies (filtering)**
* **convert analog signals to digital data**
* **transmit information over long distances**

This is why **electronics forms the backbone of modern scientific instrumentation**.


## 2. Introduction – Circuits as Models of Electrical Systems

Electronic systems exist everywhere around us:
• power supplies  
• sensors  
• computers  
• communication devices  

All of these systems are built from electrical components that interact with
each other.

Describing these systems directly using the full equations of
electromagnetism would be extremely complicated. Instead, engineers and
physicists use a simplified representation called an **electrical circuit**.

In a circuit model:
• electrical **current** describes the flow of charge  
• **voltage** describes electrical potential differences  
• components such as **resistors, capacitors and inductors** represent
physical elements of the system.

This abstraction allows us to analyze and design complex electrical systems
using relatively simple mathematical tools.

In the following sections we introduce the fundamental ideas needed to work
with circuits.




## 3. Real Systems and Ideal Circuit Models

Electrical systems that exist in the real world are always **physical objects**:
• wires  
• batteries  
• sensors  
• electronic devices  

Inside these systems, electric charge moves under the influence of
electromagnetic fields. In principle, the behaviour of such systems is
described by the **laws of electromagnetism**.

However, solving the full electromagnetic equations for practical systems
would be unnecessarily complicated. Instead, electronics uses a much simpler
representation called a **circuit model**.

In this model we describe a system using a small number of variables:
• **Voltage (V)** – electrical potential difference between two points  
• **Current (I)** – flow of electric charge through a conductor  

<u>Physical devices are represented by **ideal components**</u> such as:
• **Resistors** – oppose the flow of current  
• **Capacitors** – store energy in an electric field  
• **Inductors** – store energy in a magnetic field

<u>These elements are connected together to form an **electrical circuit**.</u>

<u>The circuit model is therefore an **approximation of the real physical
system**, but it is extremely powerful and allows us to analyze complex
devices using relatively simple mathematical tools.</u>

### The Origin of Electronic Circuits

Electronics ultimately originates from **electromagnetism**.

Electric and magnetic phenomena are described by **Maxwell’s equations**, which determine how electric and magnetic fields evolve in space and time.

However, solving Maxwell’s equations directly for every electrical device would be extremely complicated. Even simple systems would require solving field equations throughout the entire space. For this reason, electronics uses a powerful **approximation**.

#### The Lumped Circuit Approximation

For many practical systems the **physical size of the device** is much smaller than the wavelength of the signals involved. The wavelength of an electromagnetic signal is
$$ \lambda = \frac{c}{f} $$
where (c) is the speed of light and (f) is the signal frequency. If the characteristic size of the circuit (L) satisfies
$$ L \ll \lambda $$
then <u>**electromagnetic propagation effects can be neglected.**
In this situation the system can be approximated as a **lumped circuit**.</u>

By neglecting field propagation effects we obtain a model that:
* **isolate electric and magnetic filds within circuit components (lumped components e.g. resistor, capacitor, inductor etc.)**
* **Current and Voltage are defined as point-like quantities** (instead of analyzing vector field in each point of space, we utilize scalar quantities V & I for each component.
* **Using idealized elements** (passive or active).

# **Basic Elements of an Electrical Circuit**

An **electrical circuit** consists of **basic passive and active elements** that model the behavior of electrical energy. These elements have **different properties** and **mathematical models** that are used in the analysis of electrical circuits.

## **Passive Elements**
<u>**Passive elements** do not generate energy but **store, dissipate, or delay it**.</u> These include **resistors, capacitors, and inductors**.

### **Resistor ($R$)**
- **What does it model?**  
  - It models the **loss of energy** due to the resistance of conductors (e.g., copper conductors, wires, resistors).
  - It models **"electrical friction"**, where electrical energy is converted into **heat** (Joule effect).
  - Symbol: <img src="../figs/lec4/R.svg"/>
  
- The $U-I$ relationship is **Ohm’s Law:**  
  $$ U = I\,R $$
  where: $U$ is the voltage applied across the resistor, $I$ is the current flowing through it, and $R$ is the resistance in ohms ($\Omega$).

- **Properties of the Resistor:**
  - In **DC (Direct Current)**: The resistance is constant.
  - In **AC (Alternating Current)**: It does not affect the frequency or phase of the signal.
  - **Power Dissipated** by a resistor:
    $$ P = U\,I = I^2 \,R = \frac{U^2}{R} $$

### **Capacitor ($C$)**
- **What does it model?**  
  - It models the **storage of energy in an electric field** between its two plates.
  - It is used for **filtering, voltage stabilization, and storing electric charge**.
  - Symbol: <img src="../figs/lec4/C.svg"/>

- **Relationship between voltage and current:**  
  $$ I = C \,\frac{dU}{dt}$$
  where $C$ is the capacitance in farads ($F$).
  - **In DC**: In steady state ($dU/dt=0$), it behaves as an open circuit where $I=0$ (no current flows).
  - **In AC**: It provides a **capacitive reactance** (impedance) that decreases with frequency:
    $$ X_C = \frac{1}{\omega C}, \quad \text{where} \quad \omega = 2\pi f $$
  - The **Electrical Energy** stored when there is a potential difference $U$ is:
    $$ E = \frac{C\, U^2}{2} $$
  - It does not change the frequency of the signal but shifts the phase of the current relative to the voltage. For a capacitor, the current leads the voltage by $\varphi=\frac{\pi}{2}$.

### **Inductor ($L$)**
- **What does it model?**  
  - It models the **storage of energy in a magnetic field**.
  - It is used for **filtering, current stabilization, and limiting sudden changes in current**.
  - Symbol: <img src="../figs/lec4/L.svg"/>

- **Relationship between voltage and current:**  
  $$ U = L \frac{dI}{dt} $$
  where $L$ is the inductance in henries ($H$).
  - **In DC**: In steady state ($dI/dt=0$) it behaves like a normal conductor.
  - **In AC**: It provides an **inductive reactance** that increases with frequency:
    $$ X_L = \omega L $$
  - The **Magnetic Energy** stored when a current $I$ flows is:
    $$ E = \frac{L\, I^2}{2} $$
  - It does not change the frequency of the signal but shifts the phase of the current relative to the voltage. For an inductor, the current lags the voltage by $\varphi=\frac{\pi}{2}$.

## **Active Elements**

<u>**Active elements** can **generate or amplify electrical energy**.</u> These include voltage and current sources, transistors, and amplifiers.

### **Voltage Source**
- **What does it model?**  
  - It models an **element that creates a constant voltage** regardless of the current.
- **Types of voltage sources:**
  - **Ideal source**: Always provides a constant voltage $U$.
  - **Real source**: Has an **internal resistance** that affects the output voltage.
  - Symbol: <img src="../figs/lec4/Us.svg"/> (diamond symbols indicate controlled voltage sources)

### **Current Source**
- **What does it model?**  
  - It models an **element that produces a constant current**, regardless of the voltage.
- **Types of current sources:**
  - **Ideal source**: Always provides a constant current $I$.
  - **Real source**: Has an **internal resistance** that affects the output current.
  - Symbol: <img src="../figs/lec4/Is.svg"/>  (diamond symbols indicate controlled current sources)

### **Transistors (BJT & FET)**
- **What do they model?**  
  - They model **signal amplifiers**, **electronic switches**, and are used in digital logic.
  - For example, a bipolar transistor **BJT** has three terminals: **Base (B), Collector (C), Emitter (E)**.
  - A field-effect transistor **FET** has **Gate (G), Drain (D), and Source (S)**.
  - Symbol: <img src="../figs/lec4/BJT_FET.svg"/>

## **Summary**

| Element | What does it model? | Mathematical relation | Behavior in DC | Behavior in AC |
|--------|--------------------|----------------------|---------------|---------------|
| **Resistor $R$** | Energy dissipation (heat) | $U = IR$ | Acts as a current limiter | Does not affect signal phase |
| **Capacitor $C$** | Electrical energy storage | $I = C \frac{dU}{dt}$ | Blocks current after a certain time | Provides reactance $X_C = \frac{1}{\omega C}$ |
| **Inductor $L$** | Magnetic energy storage | $U = L \frac{dI}{dt}$ | After a certain time behaves like a conductor | Provides reactance $X_L = \omega L $ |
| **Voltage source** | Provides constant voltage | $U =U_0= \text{constant}$ | Always active | Can also vary with frequency |
| **Current source** | Provides constant current | $I =I_0 = \text{constant}$ | Always active | Can also vary with frequency |
| **Transistors (BJT, MOSFET)** | Controls current/voltage | Nonlinear function | Used as switch or amplifier | Can be used for amplification |

These are the **fundamental building blocks** of electrical and electronic circuits. **Resistors, capacitors, and inductors** are the main **passive elements**, while **sources and transistors** are **active elements**.

## 3. Basic Circuit Structures

Once components are introduced, they can be connected together in different
ways to form electrical circuits.

The two most fundamental connection types are:

• **Series connection**  
• **Parallel connection**

These basic structures appear in almost every electronic system.


### Series Connection

Components are said to be connected **in series** when they are placed one
after another along the same current path.

In a series circuit:

• The **same current** flows through all components  
• The **total voltage** across the circuit is the sum of the voltages across
each component

Series connections are often used when the same current must pass through
multiple elements of a circuit.


### Parallel Connection

Components are connected **in parallel** when their terminals are connected
to the same two nodes of a circuit.

In a parallel circuit:

• The **same voltage** appears across all components  
• The **total current** is the sum of the currents through each branch

Parallel connections are commonly used when multiple components must operate
under the same voltage.


| Property | Series Connection | Parallel Connection |
|--------|--------|--------|
| Current | Same current through all components | Current splits between branches |
| Voltage | Voltage divides across components | Same voltage across each branch |
| Typical use | Current flows through elements sequentially | Multiple elements operate at the same voltage |


Understanding how components are connected is essential, but we also need
tools to determine the voltages and currents inside a circuit.


### Summary

| Comonent | Series Connection | Parallel Connection |
|------------|----------------|-------------------|
| **Resistor (R)** | $R_t=R_1 + R_2 + \dots + R_n$ | $\frac{1}{R_t} = \frac{1}{R_1} + \dots + \frac{1}{R_n}$ |
| **Capacitor (C)** | $\frac{1}{C_t} = \frac{1}{C_1} + \dots + \frac{1}{C_n}$ | $C_t=C_1 + C_2 + \dots + C_n$ |
| **Inductor (L)** | $L_t=L_1 + L_2 + \dots + L_n$ | $\frac{1}{L_t} = \frac{1}{L_1} + \dots + \frac{1}{L_n} $ |
| **Voltage Source (V)** | $U_t=U_1 + U_2 + \dots + U_n$ | Must have equal voltages |
| **Current Source (I)** | Must have equal currents | $I_t=I_1 + I_2 + \dots + I_n$ |

# 4. Fundamental Laws for Circuit Analysis

Understanding how components are connected is important, but we also need
methods to determine the **voltages and currents** inside a circuit.

Circuit analysis relies on a small set of fundamental physical principles.
The most important ones are:

• **Ohm's Law**  
• **Kirchhoff's Voltage Law (KVL)**  
• **Kirchhoff's Current Law (KCL)**  

Together these laws allow us to analyze almost any electrical circuit.

---

## Ohm's Law

Ohm's law describes the relationship between **voltage, current, and resistance**
in a resistive element.

The law states:

$$
U = I R
$$

where

- $U$ is the voltage across the resistor
- $I$ is the current through the resistor
- $R$ is the resistance

Ohm's law is one of the most frequently used relations in circuit analysis.
It allows us to determine current from voltage, or voltage from current.

---

## Kirchhoff's Current Law (KCL)

Kirchhoff's Current Law expresses the **conservation of electric charge**.

It states that:

> The total current entering a node is equal to the total current leaving the node.

Mathematically:

$$
\sum I_{in} = \sum I_{out}
$$

This law follows directly from the fact that electric charge cannot accumulate
indefinitely at a point in a circuit.

Nodes where several components meet are therefore key points in circuit analysis.



## Kirchhoff's Voltage Law (KVL)

Kirchhoff's Voltage Law expresses the **conservation of energy** in a circuit loop.

It states that:

> The sum of all voltages around a closed loop must be zero.

Mathematically:

$$
\sum U = 0
$$

This means that the voltage provided by sources in a circuit must be balanced
by the voltage drops across circuit elements.


## Why These Laws Matter

Using **Ohm's law together with Kirchhoff's laws**, we can determine:

• currents through components  
• voltages at different points in the circuit  
• power dissipation in elements  

These tools form the **foundation of circuit analysis**, and they will be used
throughout this course when studying both analog and digital electronics.

# 5. Real Components and Datasheets

So far we have described circuit elements using **ideal mathematical models**.
These models are extremely useful because they simplify the analysis of
electrical circuits.

However, **real electronic components are never perfectly ideal**.

Every physical component has **limitations and additional effects**
that are not included in the simplest circuit models.

Understanding these limitations is important when designing
or building real electronic systems.



## Ideal Components vs Real Components

In circuit analysis we often assume **ideal behavior**:
- an **ideal resistor** has exactly the same resistance at all times
- an **ideal capacitor** stores energy without losses
- an **ideal inductor** stores magnetic energy without resistance

In reality, physical components deviate from these ideal models.

For example:
- resistors have **tolerance and temperature dependence**
- capacitors have **leakage currents and internal resistance**
- inductors have **resistive losses in their windings**

Because of this, engineers must always consider the **real characteristics**
of the components they use.


## Parasitic Effects

Real components also contain additional unintended electrical properties
called **parasitic elements**.

These arise from the physical structure of the component.

Typical parasitic effects include:
- **Parasitic resistance**
- **Parasitic capacitance**
- **Parasitic inductance**

For example:
- wires and component leads introduce **small inductances**
- nearby conductors create **small capacitances**
- materials introduce **resistive losses**

These effects are often negligible at low frequencies,
but they can become important in **high-frequency circuits**.

## Component Datasheets

The detailed behavior of a real component is described in its **datasheet**.

A datasheet is a technical document provided by the manufacturer that
contains important information such as:

- nominal value of the component  
- tolerance  
- maximum voltage or current  
- power limits  
- temperature range  
- frequency behavior  

When designing or building a circuit, the datasheet is the **primary source
of information** about how the component will behave in practice.



## From Models to Real Circuits

In electronics we therefore work with two complementary ideas:

1. **Ideal circuit models**  
   Used for analysis and understanding.

2. **Real components and datasheets**  
   Used when building and designing physical circuits.

**Learning to move between these two perspectives is an essential skill in electronics.**
