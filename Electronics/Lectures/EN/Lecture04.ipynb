# 1. Why Study Electronics?

<u>Electronics is the science and technology that allows us to **measure, process, transmit, and control signals**.</u>

**Almost every modern scientific instrument relies on electronics.**

Examples include:
* scientific detectors
* communication systems
* computers
* sensors
* control systems
* measurement equipment

In physics laboratories, electronics is essential because most physical phenomena must first be **converted into electrical signals** before they can be measured or analyzed.



## 1.1 Physical Quantities and Signals

Many physical systems produce quantities that vary in time.
Examples include:
* sound pressure in air
* light intensity from a star
* temperature of a material
* magnetic field near a wire
* motion of a mechanical system

To analyze these quantities we represent them as **signals**.

<u>**A signal is simply a physical quantity expressed as a function of time.**</u>

Examples:
* Voltage signal $V(t)$
* Current signal $I(t)$
* Light intensity signal $I_{\rm light}(t)$
* Sound pressure signal $p(t)$
* etc ...

Electronics allows us to **convert many physical quantities into electrical signals**, because electrical signals are easy to manipulate, amplify, and measure.



## 1.2 Sensors: Converting Physics into Electricity

Devices called **sensors** convert physical quantities into electrical signals.

Examples:

| Physical quantity | Sensor        | Electrical signal |
| ----------------- | ------------- | ----------------- |
| Temperature       | Thermistor    | resistance change |
| Light             | Photodiode    | current           |
| Sound             | Microphone    | voltage           |
| Magnetic field    | Hall sensor   | voltage           |
| Position          | Potentiometer | voltage           |

For example, a microphone converts sound pressure into a voltage signal:

$$ p(t) \rightarrow V(t) $$

Once the signal becomes electrical, electronics allows us to
* amplify it
* filter it
* record it
* transmit it
* process it



## 1.3 Why Electrical Signals?

**Electrical signals are extremely convenient because they can be easily manipulated.**
Using electronic circuits we can:
* amplify weak signals
* remove noise
* select specific frequencies
* convert analog signals to digital data
* transmit information over long distances

This is why **electronics forms the backbone of modern scientific instrumentation**.



## 1.4 Signals in Physics Experiments

In a typical physics experiment, the information flow looks like this:

```text
Physical system
       ↓
     Sensor
       ↓
Electrical signal
       ↓
Electronic processing
       ↓
Data acquisition
       ↓
Computer analysis
```

Example:
* **Muon detector**

```
Muon interaction
     ↓
Scintillator
     ↓
Photodetector
     ↓
Voltage pulse
     ↓
Amplifier
     ↓
Digitizer
     ↓
Computer
```

Almost every modern experiment follows a similar chain.


* **Smartphone sensors**
    * microphone
    * camera
    * accelerometer
    * magnetometer
    * GPS

All follow the same chain:
```text
physics → sensor → electronics → data
```



## 1.5 The Goal of Electronics

Electronics provides the tools to:
* **extract information from physical systems**
* **manipulate signals**
* **control devices**
* **build measurement instruments**

However, although electronics originates from **electromagnetism**, solving the full electromagnetic equations for every system would be extremely complicated.

Instead, electronics uses **simplified models called circuits**.

In the next section we will see how **electronic circuits arise as an approximation of electromagnetic systems**.

# 2. The Origin of Electronic Circuits

Electronics ultimately originates from **electromagnetism**.

Electric and magnetic phenomena are described by **Maxwell’s equations**, which determine how electric and magnetic fields evolve in space and time.

However, solving Maxwell’s equations directly for every electrical device would be extremely complicated. Even simple systems would require solving field equations throughout the entire space.

For this reason, electronics uses a powerful **approximation**.

## 2.1 Distributed Electromagnetic Systems

In reality, electrical systems consist of **distributed electromagnetic fields**.

Voltages and currents arise from the interaction of electric and magnetic fields in space.

In the most general case, the behavior of a system must be described using Maxwell’s equations:

$$
\begin{array}{rl} 
\nabla \times \mathbf{E} &= -\frac{\partial \mathbf{B}}{\partial t}\\
\nabla \times \mathbf{B} &= \mu_0 \mathbf{J} + \mu_0 \varepsilon_0 \frac{\partial \mathbf{E}}{\partial t}\\
\nabla \cdot \mathbf{E} &= \frac{\rho}{\varepsilon_0}\\
\nabla \cdot \mathbf{B} &= 0
\end{array}
$$

These equations describe **electromagnetic waves and field propagation**.


## 2.2 The Lumped Circuit Approximation

For many practical systems the **physical size of the device** is much smaller than the wavelength of the signals involved. The wavelength of an electromagnetic signal is
$$ \lambda = \frac{c}{f} $$
where (c) is the speed of light and (f) is the signal frequency. If the characteristic size of the circuit (L) satisfies
$$ L \ll \lambda $$
then <u>**electromagnetic propagation effects can be neglected.**
In this situation the system can be approximated as a **lumped circuit**.</u>


## 2.3 Lumped Elements

In the lumped approximation:
* electrical quantities are concentrated in **discrete components**
* voltages exist between **nodes**
* currents flow through **branches**

The distributed electromagnetic system is therefore replaced by a simplified model composed of **circuit elements** such as:
* resistors
* capacitors
* inductors

**These elements represent simplified descriptions of physical electromagnetic behavior.**



## 2.4 Why This Approximation Works

**For most laboratory electronics the frequencies involved are relatively low.**
*Example:* For a signal with frequency 
$$f = 1,\text{MHz}$$
the wavelength is
$$ \lambda = 300,\text{m} $$
Typical circuits are much smaller than this length scale, so the lumped circuit approximation is extremely accurate.

Electronic circuit on a breadboard → lumped circuit model works perfectly.

At very high frequencies (microwaves, RF engineering), however, this approximation breaks down and the full electromagnetic description becomes necessary.

Radio antenna (size comparable to wavelength) → must use electromagnetic theory.


## 2.5 From Electromagnetism to Circuits

The conceptual transition can therefore be summarized as

```text
Electromagnetic fields
        ↓ approximation
Lumped circuit model
        ↓
Nodes, voltages, currents
        ↓
Circuit elements (R, C, L)
```

Circuit theory is therefore **not a separate theory**, but a **simplified description of electromagnetism** that works when spatial dimensions are small compared to the signal wavelength.







This gives students an intuitive boundary between **electronics and electromagnetics**.
