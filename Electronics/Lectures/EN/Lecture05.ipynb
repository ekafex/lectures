# Electrical Circuit Analysis

The goal of **electrical circuit analysis** is to accurately determine the **electrical behavior of circuits**, by calculating the voltages, currents, and power dissipated in the various components.

Circuit analysis is the main tool used to **understand, design, and control** the behavior of any electrical or electronic circuit, ensuring that circuits operate correctly, safely, and efficiently.


## **Main Objectives of Circuit Analysis:**

1. **Predicting circuit behavior**
   - Precisely identifying voltages and currents at every point.
   - Verifying whether the circuit operates according to theoretical expectations.

2. **Circuit design and optimization**
   - Selecting components with optimal values.
   - Optimizing energy consumption and efficiency.

3. **Identifying problems and preventing damage**
   - Detecting extreme conditions (current, voltage).
   - Preventing overloads and component failures.

4. **Theoretical verification before physical implementation**
   - Theoretical analysis saves time, money, and effort by avoiding corrections at later stages.

5. **Diagnosing problems in existing circuits**
   - Analysis helps identify and solve problems in real circuits.


### **Main analysis methods:**

- Kirchhoff's Laws (**KVL, KCL**)
- Node Analysis
- Superposition Analysis (**Superposition**)
- Equivalent methods (**Thevenin, Norton**)


## **Method 1: Kirchhoff's Laws**

General Algorithm for Solving with KVL & KCL:

1. Identify the nodes and loops of the circuit.
2. Assign arbitrary directions for currents and polarities for voltages.
3. Apply KCL to each node (establishes the relationship between currents).
4. Apply KVL to each independent loop (establishes the relationship between voltages).
5. Apply the Uâ€“I (voltâ€“ampere) relationship for each element in the circuit.
6. Solve the resulting system of equations.

## **Method 2: Combining Elements (Series and Parallel)**

This method consists of **simplifying the circuit** by gradually combining elements (resistors, capacitors, inductors) connected in **series** or **parallel**, until a simpler circuit is obtained that can be easily analyzed.

### **Step-by-step procedure:**

#### **Step 1: Identify Combinable Elements**
- Find circuits or segments where components are clearly connected in **series** or **parallel**.

#### **Step 2: Combine Elements According to the Rules**
Apply the following rules depending on the type of element:

- **Resistors:**
  - Series: $R_{eq} = R_1 + R_2 + \dots + R_n$
  - Parallel: $\frac{1}{R_{eq}} = \frac{1}{R_1} + \frac{1}{R_2} + \dots + \frac{1}{R_n}$

- **Capacitors:**
  - Series: $\frac{1}{C_{eq}} = \frac{1}{C_1} + \frac{1}{C_2} + \dots + \frac{1}{C_n}$
  - Parallel: $C_{eq} = C_1 + C_2 + \dots + C_n$

- **Inductors:**
  - Series: $L_{eq} = L_1 + L_2 + \dots + L_n$
  - Parallel: $\frac{1}{L_{eq}} = \frac{1}{L_1} + \frac{1}{L_2} + \dots + \frac{1}{L_n}$

#### **Step 3: Progressive Circuit Simplification**
- Replace combined elements with their equivalent element.
- Repeat the process step by step, making the circuit progressively simpler.

#### **Step 4: Analyze the Simplified Circuit**
- Analyze the simplified circuit to determine the desired quantities (currents, voltages).

#### **Step 5: Return to the Original Circuit**
- Work backward using the quantities obtained in step 4 to determine the quantities for each original component.

## **Method 3: Node Analysis**

Node analysis is a concrete and systematic application of Kirchhoffâ€™s Current Law (KCL), which allows circuit analysis by determining the voltages at the nodes of the circuit.

The main goal is to determine the **node voltages**, from which the currents in every element can easily be calculated.


### **Step-by-step procedure:**

#### **Step 1: Identify the Reference Node (Ground)**  
- Choose one node as the **reference point** (ground). It is usually the lowest point or the one with the most connections. The ground symbol is <img src="../figs/lec4/GND.svg"/> and is also labeled GND.

#### **Step 2: Label the Voltages of the Other Nodes**  
- Assign unknown voltages to the remaining nodes: $V_1, V_2, \dots, V_n$.

#### **Step 3: Write KCL at Each Node**  
- Apply KCL to each node:
$$ \sum I_{\text{in}} = \sum I_{\text{out}} $$

- **Express currents as functions of node voltages:**  
If a resistor connects nodes with voltages $V_a$ and $V_b$, the current is:
$$ I = \frac{V_a - V_b}{R} $$

#### **Step 4: Solve the System of Equations**  
- Step 3 produces a system of linear equations with the unknown node voltages.
- Solve these equations using standard methods (linear algebra).

#### **Step 5: Calculate Branch Currents**  
- Once the node voltages are known (e.g., $V_a$ and $V_b$), use them to calculate the currents through each element:
$$ I_{R} = \frac{V_a - V_b}{R} $$

### **When is Node Analysis Used?**

- Particularly efficient when there are **few nodes** but many branches.
- Very suitable for complex linear circuits with several sources and passive elements.
- It is an intuitive method and well suited for computer-based numerical analysis.


#### Example 1

$$

## **Method 4: Superposition Method**

From the examples above we observe that the potential at each node $V_i$ is given as a linear combination of all sources, whether voltage or current. In other words:

$$ V_i= \sum_k \alpha_k\, U_k + \sum_l \beta_l\, I_l$$

where $U_k$ and $I_l$ are voltage and current sources, respectively.

> The Superposition Method assumes independent sources (voltage or current). The circuit output is determined by summing the responses produced by each source acting individually.

This method is closely related to **linear circuits**.

#### **What is a Linear Circuit?**

 
<div style="border:6px solid #f00; background-color:#ddd; color:#000; border-radius: 5px; padding: 10px;">

##### **Linearity of a Function (Reminder)**

$$
\text{Linearity} \Rightarrow \begin{cases} \text{Homogeneity} \\ \text{Superposition} \end{cases}
$$

Assume a function of several variables $f(x_1, x_2, ..., x_n)$.

- **Homogeneity property:**
 $$f(\alpha\, x_1, \alpha\,x_2, ..., \alpha\,x_n) = \alpha\,f(x_1, x_2, ..., x_n) \qquad \left[f(\alpha\, x) = \alpha\, f(x) \right] $$

- **Superposition property:** 
 $$f(x_1+y_1, x_2+y_2, ..., x_n+y_n) = f(x_1, x_2, ..., x_n) + f(y_1, y_2, ..., y_n) \qquad \left[ f(x+y) = f(x) + f(y) \right]$$


</div>

<br>

A circuit is called **linear** if the relationship between voltages and currents is linear. A linear circuit has simple properties:

- Superposition fully applies.
- Voltages and currents can be described by linear equations.

> **Notice that the equations appearing in the solved exercises are linear!**

## **Equivalence Method**

The equivalence method consists of replacing part of a complex electrical circuit with a simpler circuit that is fully **equivalent in terms of electrical behavior**.

<img src="../figs/lec4/equiv1.png" align="center" width=700 style="margin-left:auto; margin-right:auto"/>

*Ref: Anant Agarwal and Jeffrey Lang, course materials for 6.002 Circuits and Electronics, Spring 2007. MIT OpenCourseWare (http://ocw.mit.edu/).*


Two well-known equivalence methods are:

- **Thevenin's Theorem** (Voltage source + resistance)
- **Norton's Theorem** (Current source + resistance)

<br>

<div style="border:6px solid #f00; background-color:#ddd; color:#000; border-radius: 5px; padding: 10px;">

#### **Thevenin's Theorem**
 
Any linear circuit containing sources and resistances can be replaced by a **single voltage source $V_{TH}$** in series with a **single resistance $R_{TH}$**, viewed from the circuit terminals.

<img src="../figs/lec4/Thevenin.svg" align="center" width=200 style="margin-left:auto; margin-right:auto"/>

##### **Thevenin Parameters:**

- **Voltage $V_{TH}$**: The voltage at the open terminals of the circuit (open-circuit voltage).
- **Resistance $R_{TH}$**: The equivalent resistance seen at the terminals when the sources are deactivated.

<img src="../figs/lec4/thevenin.png" align="center" width=600 style="margin-left:auto; margin-right:auto"/> 

</div>

<br>

<div style="border:6px solid #f00; background-color:#ddd; color:#000; border-radius: 5px; padding: 10px;">

#### **Norton's Theorem**
 
Any linear circuit can be replaced by a **single current source $I_{N}$** in parallel with a **single resistance $R_{N}$**, viewed from the circuit terminals.

<img src="../figs/lec4/Norton.svg" align="center" width=200 style="margin-left:auto; margin-right:auto"/>

##### **Norton Parameters:**

- **Current $I_{N}$**: The current measured at the short-circuited terminals (short-circuit current).
- **Resistance $R_{N}$**: Has the same value as the Thevenin resistance ($R_N=R_{TH}$).

#### **Conversion between Thevenin and Norton**

The two equivalents can easily be converted from one to the other:

- From Thevenin to Norton:
$$ I_N=\frac{V_{TH}}{R_{TH}}, \quad R_N=R_{TH} $$

- From Norton to Thevenin:
$$ V_{TH}=I_N \cdot R_N, \quad R_{TH}=R_N $$


### **Algorithm for Applying the Equivalence Theorems**

1. **Identify the circuit segment to be replaced**
   - Choose a part of the circuit to replace with the Thevenin or Norton equivalent.

2. **Find $V_{TH}$ or $I_{N}$**
   - $V_{TH}$: Measure the voltage across the open terminals.
   - $I_{N}$: Measure the current across the shorted terminals.

3. **Find the resistance $R_{TH}=R_{N}$**  
   - Deactivate the sources (voltage sources become short circuits, current sources become open circuits).
   - Measure the equivalent resistance seen from the terminals.

4. **Replace the original segment with the equivalent**
   - Use the equivalent circuit (Thevenin or Norton) to simplify analysis.


### **Connection with the Superposition Method**

In linear circuits, Thevenin and Norton theorems work very well with the superposition method. Superposition states:

> In a linear circuit with multiple sources, the total effect at any point is the algebraic sum of the effects produced by each source acting individually, with the other sources turned off.

### **Combined Algorithm (Superposition + Thevenin/Norton):**

- **Step 1**: Analyze the effect of each source individually (turn off the others).
- **Step 2**: For each individual analysis, use the Thevenin or Norton equivalent to simplify the circuit.
- **Step 3**: Calculate the currents and voltages in each case.
- **Step 4**: Sum the results (superposition) to obtain the final solution.


### **Short Summary:**

- **Linear circuit**: Can be replaced by simple equivalents (Thevenin/Norton).
- **Thevenin**: Voltage source + series resistance.
- **Norton**: Current source + parallel resistance.
- **Superposition**: Works very well together with equivalence methods for simple and effective circuit analysis.
