# Nga Elektromagnetizmi te Qarqet Elektronike


Literatura për këtë kapitull: Lexoni kapitullin e parë të [***Foundations of Analog and Digital Electronic Circuits***, A. Agarwal & J. Lang](https://neurophysics.ucsd.edu/courses/physics_120/Agarwal%20and%20Lang%20(2005)%20Foundations%20of%20Analog%20and%20Digital.pdf)


## 1. Pse të Studiojmë Elektronikën?

<u>Elektronika është shkenca dhe teknologjia që na lejon të **masim, përpunojmë, transmetojmë dhe kontrollojmë sinjale**.</u>

**Pothuajse çdo instrument shkencor modern mbështetet në elektronikë.**

Shembuj përfshijnë:
* detektorë shkencorë
* sisteme komunikimi
* kompjuterë
* sensorë
* sisteme kontrolli
* pajisje matëse

Në laboratorët e fizikës, elektronika është thelbësore sepse shumica e fenomeneve fizike duhet së pari të **konvertohen në sinjale elektrike** përpara se të mund të maten ose analizohen.

Shembuj përfshijnë:
* presioni i zërit në ajër
* intensiteti i dritës nga një yll
* temperatura e një materiali
* fusha magnetike pranë një përçuesi
* lëvizja e një sistemi mekanik

Për të analizuar këto madhësi, ne i paraqesim ato si **sinjale**.

<u>**Një sinjal është thjesht një madhësi fizike e shprehur si funksion i kohës.**</u>

Shembuj:
* Sinjali i tensionit $V(t)$
* Sinjali i rrymës $I(t)$
* Sinjali i intensitetit të dritës $I_{\rm light}(t)$
* Sinjali i presionit të zërit $p(t)$
* etj ...

Elektronika na lejon të **konvertojmë shumë madhësi fizike në sinjale elektrike**, sepse sinjalet elektrike janë të lehta për t’u manipuluar, përforcuar dhe matur.


### Pse Sinjale Elektrike?

**Sinjalet elektrike janë shumë të përshtatshme sepse mund të manipulohen lehtësisht.**

Duke përdorur qarqe elektronike ne mund të:
* **përforcojmë sinjale të dobëta**
* **eliminojmë zhurmën**
* **përzgjedhim frekuenca specifike (filtrim)**
* **konvertojmë sinjale analoge në të dhëna digjitale**
* **transmetojmë informacion në distanca të gjata**

Kjo është arsyeja pse **elektronika përbën bazën e instrumentacionit shkencor modern**.

## 2. Hyrje – Qarqet si Modele të Sistemeve Elektrike

Sistemet elektronike gjenden kudo rreth nesh:
• burime energjie  
• sensorë  
• kompjuterë  
• pajisje komunikimi  

Të gjitha këto sisteme ndërtohen nga komponentë elektrikë që ndërveprojnë me
njëri-tjetrin.

Përshkrimi i këtyre sistemeve drejtpërdrejt duke përdorur ekuacionet e plota të
elektromagnetizmit do të ishte shumë i ndërlikuar. Në vend të kësaj,
inxhinierët dhe fizikantët përdorin një paraqitje të thjeshtuar të quajtur
**qark elektrik**.

Në modelin e qarkut:

• **rryma elektrike** përshkruan rrjedhën e ngarkesës  
• **tensioni** përshkruan diferencat e potencialit elektrik  
• komponentë si **rezistencat, kondensatorët dhe bobinat** përfaqësojnë
elementë fizikë të sistemit.

Ky abstraksion na lejon të analizojmë dhe projektojmë sisteme elektrike
komplekse duke përdorur mjete matematikore relativisht të thjeshta.

Në seksionet në vijim do të prezantojmë idetë themelore të nevojshme për të
punuar me qarqe elektrike.

## 3. Sistemet Reale dhe Modelet Ideale të Qarqeve

Sistemet elektrike që ekzistojnë në botën reale janë gjithmonë **objekte fizike**:
• tela  
• bateri  
• sensorë  
• pajisje elektronike  

Brenda këtyre sistemeve, ngarkesa elektrike lëviz nën ndikimin e
fushave elektromagnetike. Në parim, sjellja e këtyre sistemeve
përshkruhet nga **ligjet e elektromagnetizmit**.

Megjithatë, zgjidhja e ekuacioneve të plota elektromagnetike për sisteme
praktike do të ishte tepër e ndërlikuar. Në vend të kësaj, elektronika
përdor një paraqitje shumë më të thjeshtë të quajtur **model qarku**.

Në këtë model ne përshkruajmë një sistem duke përdorur një numër të vogël
madhësish:

• **Tensioni (V)** – diferenca e potencialit elektrik midis dy pikave  
• **Rryma (I)** – rrjedha e ngarkesës elektrike nëpër një përçues  

<u>Pajisjet fizike përfaqësohen nga **komponentë idealë**</u> si:

• **Rezistencat** – kundërshtojnë rrjedhën e rrymës  
• **Kondensatorët** – ruajnë energji në një fushë elektrike  
• **Bobinat** – ruajnë energji në një fushë magnetike  

<u>Këta elementë lidhen së bashku për të formuar një **qark elektrik**.</u>

<u>Prandaj modeli i qarkut është një **aproksim i sistemit fizik real**,
por është jashtëzakonisht i fuqishëm dhe na lejon të analizojmë pajisje
komplekse duke përdorur mjete matematikore relativisht të thjeshta.</u>

### Origjina e Qarqeve Elektronike

Elektronika në thelb buron nga **elektromagnetizmi**.

Fenomenet elektrike dhe magnetike përshkruhen nga **ekuacionet e Maxwell-it**, të cilat përcaktojnë se si fushat elektrike dhe magnetike zhvillohen në hapësirë dhe kohë.

Megjithatë, zgjidhja e drejtpërdrejtë e ekuacioneve të Maxwell-it për çdo pajisje elektrike do të ishte jashtëzakonisht e ndërlikuar. Edhe sistemet e thjeshta do të kërkonin zgjidhjen e ekuacioneve të fushës në të gjithë hapësirën. Për këtë arsye, elektronika përdor një **aproksim të fuqishëm**.

#### Aproksimi i Qarkut të Përqendruar (Lumped Circuit Approximation)

Për shumë sisteme praktike **madhësia fizike e pajisjes** është shumë më e vogël se gjatësia valore e sinjaleve të përfshira. Gjatësia valore e një sinjali elektromagnetik është

$$ \lambda = \frac{c}{f} $$

ku (c) është shpejtësia e dritës dhe (f) është frekuenca e sinjalit. Nëse madhësia karakteristike e qarkut (L) plotëson kushtin

$$ L \ll \lambda $$

atëherë <u>**efektet e përhapjes elektromagnetike mund të neglizhohen.  
Në këtë situatë sistemi mund të aproksimohet si një **qark i përqendruar (lumped circuit)**.</u>

Duke neglizhuar efektet e përhapjes së fushës ne përfitojmë një model që:

* **izolon fushat elektrike dhe magnetike brenda komponentëve të qarkut** (komponentë të përqendruar si rezistencë, kondensator, bobinë etj.)
* **Rryma dhe tensioni përkufizohen si madhësi të përqendruara në pika** (në vend që të analizojmë fusha vektoriale në çdo pikë të hapësirës, përdorim madhësitë skalare V dhe I për çdo komponent)
* **përdor elementë të idealizuar** (pasivë ose aktivë)

# **Elementët Bazë të një Qarku Elektrik**

Një **qark elektrik** përbëhet nga **elementë bazë pasivë dhe aktivë** që modelojnë sjelljen e energjisë elektrike. Këta elementë kanë **veti të ndryshme** dhe **modele matematikore** që përdoren në analizën e qarqeve elektrike.

## **Elementët Pasivë**

<u>**Elementët pasivë** nuk gjenerojnë energji por **e ruajnë, e shpërndajnë ose e vonojnë atë**.</u> Këta përfshijnë **rezistencat, kondensatorët dhe bobinat**.

### **Rezistenca ($R$)**

- **Çfarë modelon?**  
  - Modelon **humbjen e energjisë** për shkak të rezistencës së përçuesve (p.sh., përçues bakri, tela, rezistenca).
  - Modelon **“fërkimin elektrik”**, ku energjia elektrike shndërrohet në **nxehtësi** (efekti Joule).
  - Simboli: <img src="../figs/lec4/R.svg"/>

- Marrëdhënia $U-I$ është **Ligji i Ohm-it:**

$$ U = I\,R $$

ku: $U$ është tensioni mbi rezistencën, $I$ është rryma që kalon nëpër të dhe $R$ është rezistenca në ohm ($\Omega$).

- **Vetitë e Rezistencës:**
  - Në **DC (Rrymë të vazhduar)**: Rezistenca është konstante.
  - Në **AC (Rrymë alternative)**: Nuk ndikon në frekuencën apo fazën e sinjalit.
  - **Fuqia e çliruar** nga një rezistencë:

$$ P = U\,I = I^2 \,R = \frac{U^2}{R} $$

### **Kondensatori ($C$)**

- **Çfarë modelon?**  
  - Modelon **ruajtjen e energjisë në fushën elektrike** midis dy pllakave të tij.
  - Përdoret për **filtrim, stabilizim të tensionit dhe ruajtje të ngarkesës elektrike**.
  - Simboli: <img src="../figs/lec4/C.svg"/>

- **Lidhja ndërmjet tensionit dhe rrymës:**

$$ I = C \,\frac{dU}{dt} $$

ku $C$ është kapaciteti në farad ($F$).

- **Në DC**: Në gjendje stacionare ($dU/dt=0$), vepron si qark i hapur ku $I=0$.
- **Në AC**: Ofron një **reaktancë kapacitive** që zvogëlohet me frekuencën:

$$ X_C = \frac{1}{\omega C}, \quad \text{ku} \quad \omega = 2\pi f $$

- **Energjia elektrike** që ruan kur ka diferencë potenciali $U$ është:

$$ E = \frac{C\, U^2}{2} $$

- Nuk ndryshon frekuencën e sinjalit, por sfazon rrymën nga tensioni. Për kondensatorin, rryma është përpara në fazë nga tensioni me $\varphi=\frac{\pi}{2}$.

### **Bobina ($L$)**

- **Çfarë modelon?**  
  - Modelon **ruajtjen e energjisë në një fushë magnetike**.
  - Përdoret për **filtrim, stabilizim të rrymës dhe kufizim të ndryshimeve të papritura në rrymë**.
  - Simboli: <img src="../figs/lec4/L.svg"/>

- **Lidhja ndërmjet tensionit dhe rrymës:**

$$ U = L \frac{dI}{dt} $$

ku $L$ është induktanca në henri ($H$).

- **Në DC**: Në gjendje stacionare ($dI/dt=0$) vepron si përçues i zakonshëm.
- **Në AC**: Ofron një **reaktancë induktive** që rritet me frekuencën:

$$ X_L = \omega L $$

- **Energjia magnetike** që ruan kur kalon rryma $I$ është:

$$ E = \frac{L\, I^2}{2} $$

- Nuk ndryshon frekuencën e sinjalit, por sfazon rrymën nga tensioni. Për bobinën, rryma është në vonesë faze me $\varphi=\frac{\pi}{2}$ ndaj tensionit.

## **Elementët Aktivë**

<u>**Elementët aktivë** mund të **gjenerojnë ose amplifikojnë energji elektrike**.</u> Këta përfshijnë burimet e tensionit dhe rrymës, transistorët dhe amplifikatorët.

### **Burimi i Tensionit**

- **Çfarë modelon?**  
  - Modelon një **element që krijon tension konstant** pavarësisht rrymës.

- **Llojet e burimeve të tensionit:**
  - **Burimi ideal**: Gjithmonë ofron një tension konstant $U$.
  - **Burimi real**: Ka një **rezistencë të brendshme** që ndikon në tensionin dalës.

  - Simboli: <img src="../figs/lec4/Us.svg"/> (simboli me romb tregon burime të komanduar nga tensioni)

### **Burimi i Rrymës**

- **Çfarë modelon?**  
  - Modelon një **element që krijon një rrymë konstante**, pavarësisht tensionit.

- **Llojet e burimeve të rrymës:**
  - **Burimi ideal**: Gjithmonë ofron një rrymë konstante $I$.
  - **Burimi real**: Ka një **rezistencë të brendshme** që ndikon në rrymën dalëse.

  - Simboli: <img src="../figs/lec4/Is.svg"/> (simboli me romb tregon burime të komanduar nga rryma)

### **Transistorët (BJT & FET)**

- **Çfarë modelojnë?**  
  - Modelojnë **përforcues sinjali**, **çelësa elektronikë**, dhe përdoren në logjikën digjitale.
  - Për shembull, një transistor bipolar **BJT** ka tre terminale: **Baza (B), Kolektori (C), Emiteri (E)**.
  - Një transistor me efekt fushe **FET** ka **Gate (G), Drain (D) dhe Source (S)**.
  - Simboli: <img src="../figs/lec4/BJT_FET.svg"/>

## 3. Strukturat Bazë të Qarqeve

Pasi të jenë prezantuar komponentët, ata mund të lidhen së bashku në mënyra
të ndryshme për të formuar qarqe elektrike.

Dy mënyrat më themelore të lidhjes janë:

• **Lidhja në seri**  
• **Lidhja në paralel**

Këto struktura bazë shfaqen pothuajse në çdo sistem elektronik.


### Lidhja në Seri

Komponentët thuhet se janë të lidhur **në seri** kur vendosen njëri pas
tjetrit përgjatë së njëjtës rrugë të rrymës.

Në një qark në seri:

• **E njëjta rrymë** kalon nëpër të gjithë komponentët  
• **Tensioni total** në qark është shuma e tensioneve në secilin komponent

Lidhjet në seri përdoren shpesh kur e njëjta rrymë duhet të kalojë nëpër
disa elementë të një qarku.


### Lidhja në Paralel

Komponentët janë të lidhur **në paralel** kur terminalet e tyre janë të
lidhur me të njëjtat dy nyje të qarkut.

Në një qark paralel:

• **I njëjti tension** shfaqet në të gjithë komponentët  
• **Rryma totale** është shuma e rrymave që kalojnë në secilën degë

Lidhjet në paralel përdoren zakonisht kur disa komponentë duhet të
punojnë nën të njëjtin tension.


| Vetia | Lidhja në Seri | Lidhja në Paralel |
|--------|--------|--------|
| Rryma | E njëjta rrymë kalon nëpër të gjithë komponentët | Rryma ndahet midis degëve |
| Tensioni | Tensioni ndahet midis komponentëve | I njëjti tension në secilën degë |
| Përdorimi tipik | Rryma kalon në mënyrë sekuenciale në elementë | Shumë elementë punojnë në të njëjtin tension |


Të kuptuarit se si lidhen komponentët është thelbësor, por gjithashtu na
duhen mjete për të përcaktuar tensionet dhe rrymat brenda një qarku.


### Përmbledhje

| Komponenti | Lidhja në Seri | Lidhja në Paralel |
|------------|----------------|-------------------|
| **Rezistenca (R)** | $R_t=R_1 + R_2 + \dots + R_n$ | $\frac{1}{R_t} = \frac{1}{R_1} + \dots + \frac{1}{R_n}$ |
| **Kondensatori (C)** | $\frac{1}{C_t} = \frac{1}{C_1} + \dots + \frac{1}{C_n}$ | $C_t=C_1 + C_2 + \dots + C_n$ |
| **Bobina (L)** | $L_t=L_1 + L_2 + \dots + L_n$ | $\frac{1}{L_t} = \frac{1}{L_1} + \dots + \frac{1}{L_n} $ |
| **Burimi i tensionit (V)** | $U_t=U_1 + U_2 + \dots + U_n$ | Duhet të kenë të njëjtin tension |
| **Burimi i rrymës (I)** | Duhet të kenë të njëjtën rrymë | $I_t=I_1 + I_2 + \dots + I_n$ |

```text

Sistemet fizike reale
        ↓
Abstraksioni i qarkut
        ↓
Komponentët e qarkut
        ↓
Si lidhen komponentët
(seri / paralel)
        ↓
Si analizohen qarqet
(Ohm + Kirchhoff)


<br>
<br>

# 4. Ligjet Themelore për Analizën e Qarqeve

Të kuptuarit se si lidhen komponentët është e rëndësishme, por gjithashtu na
duhen metoda për të përcaktuar **tensionet dhe rrymat** brenda një qarku.

Analiza e qarqeve mbështetet në një numër të vogël parimesh fizike
themelore. Më të rëndësishmet janë:

• **Ligji i Ohm-it**  
• **Ligji i Kirchhoff-it për Tensionin (KVL)**  
• **Ligji i Kirchhoff-it për Rrymën (KCL)**  

Së bashku, këto ligje na lejojnë të analizojmë pothuajse çdo qark elektrik.

---

## Ligji i Ohm-it

Ligji i Ohm-it përshkruan marrëdhënien midis **tensionit, rrymës dhe
rezistencës** në një element rezistiv.

Ligji shprehet si:

$$
U = I R
$$

ku

- $U$ është tensioni në skajet e rezistencës  
- $I$ është rryma që kalon nëpër rezistencë  
- $R$ është rezistenca

Ligji i Ohm-it është një nga marrëdhëniet më të përdorura në analizën e
qarqeve. Ai na lejon të përcaktojmë rrymën nga tensioni, ose tensionin nga
rryma.

---

## Ligji i Kirchhoff-it për Rrymën (KCL)

Ligji i Kirchhoff-it për rrymën shpreh **ruajtjen e ngarkesës elektrike**.

Ai thotë se:

> Shuma e të gjitha rrymave që hyjnë në një nyje është e barabartë me
> shumën e rrymave që dalin nga ajo nyje.

Matematikisht:

$$
\sum I_{in} = \sum I_{out}
$$

Ky ligj rrjedh drejtpërdrejt nga fakti se ngarkesa elektrike nuk mund të
grumbullohet pafundësisht në një pikë të qarkut.

Nyjet ku takohen disa komponentë janë prandaj pika kyçe në analizën e
qarqeve.


## Ligji i Kirchhoff-it për Tensionin (KVL)

Ligji i Kirchhoff-it për tensionin shpreh **ruajtjen e energjisë** në një
lak të qarkut.

Ai thotë se:

> Shuma e të gjitha tensioneve përgjatë një laku të mbyllur duhet të jetë
> zero.

Matematikisht:

$$
\sum U = 0
$$

Kjo do të thotë se tensioni i siguruar nga burimet në një qark duhet të
balancohet nga rëniet e tensionit në elementët e qarkut.


## Pse Janë të Rëndësishme Këto Ligje

Duke përdorur **ligjin e Ohm-it së bashku me ligjet e Kirchhoff-it**, ne
mund të përcaktojmë:

• rrymat që kalojnë nëpër komponentë  
• tensionet në pika të ndryshme të qarkut  
• fuqinë e shpërndarë në elementë  

Këto mjete përbëjnë **bazën e analizës së qarqeve**, dhe do të përdoren
gjatë gjithë këtij kursi kur të studiojmë si elektronikën analoge ashtu
edhe atë digjitale.

<br>
<br>

# 5. Komponentët Realë dhe Datasheet-et

Deri tani kemi përshkruar elementët e qarkut duke përdorur **modele matematikore ideale**.
Këto modele janë shumë të dobishme sepse thjeshtojnë analizën e
qarqeve elektrike.

Megjithatë, **komponentët realë elektronikë nuk janë kurrë plotësisht idealë**.

Çdo komponent fizik ka **kufizime dhe efekte shtesë**
që nuk përfshihen në modelet më të thjeshta të qarqeve.

Kuptimi i këtyre kufizimeve është i rëndësishëm kur projektojmë
ose ndërtojmë sisteme reale elektronike.



## Komponentët Idealë kundrejt Komponentëve Realë

Në analizën e qarqeve shpesh supozojmë **sjellje ideale**:

- një **rezistencë ideale** ka saktësisht të njëjtën rezistencë në çdo moment
- një **kondensator ideal** ruan energjinë pa humbje
- një **induktor ideal** ruan energjinë magnetike pa rezistencë

Në realitet, komponentët fizikë devijojnë nga këto modele ideale.

Për shembull:

- rezistencat kanë **tolerancë dhe varësi nga temperatura**
- kondensatorët kanë **rryma rrjedhjeje dhe rezistencë të brendshme**
- bobinat kanë **humbje rezistive në mbështjelljet e tyre**

Për këtë arsye, inxhinierët duhet gjithmonë të marrin në konsideratë
**karakteristikat reale** të komponentëve që përdorin.


## Efektet Parazite

Komponentët realë përmbajnë gjithashtu veti elektrike shtesë të
paqëllimshme të quajtura **elemente parazite**.

Këto lindin nga struktura fizike e komponentit.

Efektet tipike parazite përfshijnë:

- **Rezistencë parazite**
- **Kapacitet parazit**
- **Induktancë parazite**

Për shembull:

- telat dhe terminalet e komponentëve krijojnë **induktanca të vogla**
- përçuesit e afërt krijojnë **kapacitete të vogla**
- materialet krijojnë **humbje rezistive**

Këto efekte shpesh janë të papërfillshme në frekuenca të ulëta,
por mund të bëhen të rëndësishme në **qarqet me frekuencë të lartë**.

<table>
 <tr>
    <td> <img src="../figs/lec4/real_circuit.jpeg" align="center" width="500px"/> </td>
    <td> <img src="../figs/lec4/real_circuit2.png" align="center" width="450px"/> </td>
 </tr>
</table>



## Datasheet-et e Komponentëve

Sjellja e detajuar e një komponenti real përshkruhet në **datasheet-in** e tij.

Një datasheet është një dokument teknik i ofruar nga prodhuesi që
përmban informacione të rëndësishme si:

- vlera nominale e komponentit  
- toleranca  
- tensioni ose rryma maksimale  
- kufijtë e fuqisë  
- diapazoni i temperaturës  
- sjellja në frekuencë  

Kur projektojmë ose ndërtojmë një qark, datasheet-i është **burimi
kryesor i informacionit** për mënyrën se si komponenti do të sillet
në praktikë.



## Nga Modelet te Qarqet Reale

Në elektronikë ne punojmë me dy ide plotësuese:

1. **Modelet ideale të qarqeve**  
   Përdoren për analizë dhe për të kuptuar sistemin.

2. **Komponentët realë dhe datasheet-et**  
   Përdoren kur ndërtojmë dhe projektojmë qarqe fizike.

**Të mësuarit për të kaluar midis këtyre dy perspektivave është një aftësi thelbësore në elektronikë.**

<br>
<br>

# Përmbledhje e Leksionit — Harta Konceptuale

Në këtë leksion prezantuam **idetë themelore të nevojshme për të kuptuar qarqet elektrike**.

Konceptet kryesore ndjekin një zinxhir të thjeshtë logjik:

<div class="alert alert-info">

```text

Sistemet elektrike reale
(bateri, tela, pajisje elektronike)
        ↓
Abstraksioni i qarkut
(tensioni, rryma, komponentë idealë)
        ↓
Strukturat bazë të qarqeve
(lidhjet në seri dhe paralel)
        ↓
Mjetet për analizën e qarqeve
(Ligji i Ohm-it, Ligjet e Kirchhoff-it)
        ↓
Komponentët realë
(devijime nga sjellja ideale)
        ↓
Datasheet-et
(përshkrimi i kufizimeve reale të komponentëve)

```
</div>

Idetë Kryesore për t’u Mbajtur Mend

* Qarqet elektrike janë modele të thjeshtuara të sistemeve fizike reale.
* Komponentë si rezistencat, kondensatorët dhe bobinat përfaqësojnë sjellje të idealizuara që thjeshtojnë analizën.
* Strukturat e qarqeve si lidhjet në seri dhe paralel përcaktojnë se si shpërndahen tensionet dhe rrymat.
* Ligji i Ohm-it dhe ligjet e Kirchhoff-it ofrojnë mjetet bazë për analizën e qarqeve elektrike.
* Komponentët realë nuk janë plotësisht idealë dhe sjellja e tyre përshkruhet në datasheet-et përkatëse.

## **Ushtrime Experimentale**

## **Ushtrimi 1: Matja dhe analiza e rezistencës së vërtetë**

### **Qëllimi:**
- Të demonstrohet efekti i **rezistencës parazite** në telat dhe komponentët fizikë.

### **Pajisjet laboratorike:**
- Multimetër digjital (ose analog)
- Rezistenca nominale (p.sh. 10 Ω, 100 Ω, 1 kΩ)
- Tela me gjatësi të ndryshme dhe breadboard.

### **Procedura:**
- Matni rezistencën nominale në laborator dhe krahasojeni atë me vlerën e përcaktuar nga ngjyra.
- Matni rezistencën e telave me gjatësi të ndryshme (p.sh., 10 cm, 50 cm, 1 m).

### **Analiza dhe raportimi:**
- Krahaso rezistencën e matur me atë teorike dhe diskuto efektet e rezistencës parazite.

<br>

## **Ushtrimi 2: Efekti i Kapacitetit Parazit në qarqe me frekuenca të larta**

### **Qëllimi:**
- Të demonstrohet efekti i kapacitetit parazit në sjelljen e rezistencave dhe qarqeve në frekuencë të lartë.

### **Pajisjet laboratorike:**
- Osciloskop
- Gjenerator funksionesh (për frekuenca > 1 MHz)
- Rezistenca të mëdha (p.sh., 100 kΩ, 1 MΩ)
- Breadboard dhe tela lidhës

### **Procedura:**
- Aplikoni sinjale sinusoidale me frekuenca në intervalin 1 kHz – 10 MHz në rezistencën e madhe.
- Matni tensionin dhe fazën mes tensionit dhe rrymës për rezistenca të ndryshme dhe frekuenca të ndryshme.

### **Analiza dhe raportimi:**
- Krahaso rezultatet eksperimentale me parashikimin teorik duke përfshirë kapacitetin parazit (~0.2–1 pF).
- Diskutoni pse efekti bëhet i dukshëm në frekuenca të larta.

<br>

## **Ushtrimi 3: Induktanca Parazite dhe Impedanca në Qarqe RLC**

### **Qëllimi:**
- Të shfaqet efekti i induktancës parazite në një rezistencë të zakonshme aksiale në qarqe me frekuencë të lartë.

### **Pajisjet laboratorike:**
- Osciloskop
- Gjenerator funksionesh (0.1–10 MHz)
- Rezistenca standarde (p.sh., 100 Ω–1 kΩ)
- Kondensatorë standard (p.sh., 10 nF, 100 nF)
- Breadboard dhe tel lidhës

### **Procedura:**
- Ndërtoni një qark të thjeshtë RC dhe aplikoni sinjale me frekuencë të lartë.
- Matni tensionin dhe fazën në rezistencë dhe kondensator.
- Përsëritni matjet për rezistenca dhe kondensatorë të ndryshëm.

### **Analiza dhe raportimi:**
- Diskuto diferencat e matjeve nga modeli ideal teorik RC, duke përfshirë induktancën parazite (~10–20 nH).
