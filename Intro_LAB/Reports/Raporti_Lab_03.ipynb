## 1. Të dhënat bruto

Vendosni të dhënat tuaja këtu.  
**Njësia** duhet të jetë e qartë (mm, s, g, etj.).

> Shembulli më poshtë është ilustrues. Zëvendësojeni me të dhënat tuaja.


```python
import numpy as np
import matplotlib.pyplot as plt

# TODO: Zëvendësoni me të dhënat tuaja (p.sh. 20 matje)
data = np.array([
    10.1, 10.2, 9.9, 10.0, 10.3,
    10.1, 10.2, 10.0, 9.8, 10.1,
    10.2, 10.1, 10.0, 9.9, 10.2
])

unit = "mm"  # TODO: vendosni njësinë tuaj
data
```

## 2. Statistikat bazë

Përdorni:
- `np.mean(data)` për mesataren
- `np.std(data, ddof=1)` për devijimin standard të kampionit

Formulat:

$$
\bar{x} = \frac{1}{N}\sum_{i=1}^{N}x_i
$$

$$
s = \sqrt{\frac{1}{N-1}\sum_{i=1}^{N}(x_i-\bar{x})^2}
$$

Gabimi i mesatares:

$$
\sigma_{\bar{x}} = \frac{s}{\sqrt{N}}
$$


```python
N = len(data)
xbar = np.mean(data)
s = np.std(data, ddof=1)
sem = s / np.sqrt(N)

N, xbar, s, sem
```

```python
# Raportimi i rezultatit (kontrolloni shifrat domethënëse)
print(f"Rezultati: {xbar:.3f} ± {sem:.3f} {unit}")
```

## 3. Histogrami

Histogrami tregon shpërndarjen e matjeve.

Kërkesa minimale për grafik:
- etiketa boshtesh + njësi
- titull
- grid


```python
plt.hist(data, bins=6)
plt.xlabel(f"Vlera ({unit})")
plt.ylabel("Frekuenca")
plt.title("Histogrami i matjeve")
plt.grid(True)
plt.show()
```

## 4. Grafiku i matjeve sipas rendit

Ky grafik ndihmon të identifikoni:
- prirje (drift) gjatë matjeve
- matje të dyshimta (outliers)


```python
plt.plot(data, "o-")
plt.xlabel("Numri i matjes")
plt.ylabel(f"Vlera ({unit})")
plt.title("Matjet sipas rendit")
plt.grid(True)
plt.show()
```
