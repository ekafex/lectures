## 6. NumPy për punë numerike

Në analizë shkencore përdoret shpesh biblioteka **NumPy**.
Ajo lejon punë të përshtatshme me vektorë dhe matrica.


````python
import numpy as np

time = np.array([0, 1, 2, 3, 4])
position = np.array([0, 5, 20, 45, 80])

print("Koha:", time)
print("Pozicioni:", position)
````

## 7. Grafik me Matplotlib

Grafikët janë thelbësorë në raportet laboratorike.
Gjithmonë vendosni:

- emërtim të boshteve
- titull
- grid kur është e nevojshme


````python
import matplotlib.pyplot as plt

plt.plot(time, position, 'o-')
plt.xlabel("Koha (s)")
plt.ylabel("Pozicioni (m)")
plt.title("Shembull i një eksperimenti lëvizjeje")
plt.grid()
plt.show()
````

## 8. Një model fizik i thjeshtë

Supozojmë se po studiojmë lëvizje me përshpejtim konstant.
Modeli teorik mund të shkruhet si:

$$
x(t) = x_0 + v_0 t + \frac{1}{2} a t^2
$$

Le të ndërtojmë të dhëna teorike me këtë model.


````python
t = np.linspace(0, 4, 100)
x0 = 0
v0 = 0
a = 10

x = x0 + v0*t + 0.5*a*t**2

plt.plot(t, x)
plt.xlabel("Koha (s)")
plt.ylabel("Pozicioni (m)")
plt.title("Model teorik për lëvizje me përshpejtim konstant")
plt.grid()
plt.show()
````

## 9. Fitting linear i thjeshtë

Shpesh në laborator duam të përshtatim një model me të dhënat.
Këtu po japim një shembull shumë të thjeshtë linear.


````python
x_data = np.array([1, 2, 3, 4, 5])
y_data = np.array([2.1, 4.0, 6.2, 7.9, 10.1])

coeff = np.polyfit(x_data, y_data, 1)
slope = coeff[0]
intercept = coeff[1]

y_fit = slope * x_data + intercept

print("Pjerrësia =", slope)
print("Prerja =", intercept)

plt.scatter(x_data, y_data, label="Të dhënat")
plt.plot(x_data, y_fit, label="Fit linear")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Shembull i një fit-i linear")
plt.legend()
plt.grid()
plt.show()
````
